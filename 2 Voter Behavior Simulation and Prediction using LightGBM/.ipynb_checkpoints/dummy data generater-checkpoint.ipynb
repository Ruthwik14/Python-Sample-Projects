```python
import pandas as pd
import random
from faker import Faker
from collections import defaultdict
from datetime import datetime

fake = Faker()

# US state population (not used directly, but kept for realism)
us_states_population = {
    "California": 39500000, "Texas": 29000000, "Florida": 21500000, "New York": 19400000,
    "Pennsylvania": 12800000, "Illinois": 12600000, "Ohio": 11700000, "Georgia": 10700000,
    "North Carolina": 10400000, "Michigan": 10000000, "New Jersey": 8900000, "Virginia": 8600000,
    "Washington": 7600000, "Arizona": 7300000, "Massachusetts": 7000000, "Tennessee": 6900000,
    "Indiana": 6800000, "Missouri": 6100000, "Maryland": 6100000, "Wisconsin": 5900000,
    "Colorado": 5800000, "Minnesota": 5700000, "South Carolina": 5200000, "Alabama": 5000000,
    "Louisiana": 4600000, "Kentucky": 4500000, "Oregon": 4300000, "Oklahoma": 4000000,
    "Connecticut": 3600000, "Utah": 3300000, "Iowa": 3200000, "Nevada": 3100000,
    "Arkansas": 3000000, "Mississippi": 2900000, "Kansas": 2900000, "New Mexico": 2100000,
    "Nebraska": 2000000, "Idaho": 1800000, "West Virginia": 1800000, "Hawaii": 1400000,
    "New Hampshire": 1400000, "Maine": 1300000, "Rhode Island": 1100000, "Montana": 1100000,
    "Delaware": 1000000, "South Dakota": 900000, "North Dakota": 800000,
    "Alaska": 730000, "Vermont": 640000, "District of Columbia": 710000, "Wyoming": 580000
}

unique_agents = [fake.unique.user_name() for _ in range(25)]

ages = ["18-24", "25-34", "35-44", "45-59", "60+"]
genders = ["Male", "Female"]
ethnicities = [
    "White (Non-Hispanic)", "Black or African American", "Native American or Alaska Native",
    "Asian", "Pacific Islander", "Mixed Race", "Hispanic or Latino (any race)"
]

# Full survey field pool
survey_fields = [
    "biggest_issues", "state_govt_performance", "federal_govt_performance", "opposition_performance",
    "election_winner_prediction", "voting_preference", "best_govt_choice", "education_level", "occupation",
    "monthly_household_income", "trust_in_media", "top_media_source", "immigration_policy_opinion",
    "climate_change_priority", "gun_control_support", "abortion_rights_opinion", "party_affiliation"
]

all_state_data_distributions = {
    # Alabama
    "Alabama": {
        "biggest_issues": {"Healthcare": 0.20, "Unemployment": 0.25, "Education": 0.15, "Inflation": 0.20, "Housing": 0.05, "Crime": 0.15},
        "state_govt_performance": {"Very Good": 0.15, "Good": 0.35, "Average": 0.30, "Bad": 0.10, "Very Bad": 0.10}, # Republican-led, lean positive
        "federal_govt_performance": {"Very Good": 0.05, "Good": 0.15, "Average": 0.20, "Bad": 0.30, "Very Bad": 0.30}, # Red state, lean negative for federal
        "opposition_performance": {"Very Good": 0.15, "Good": 0.30, "Average": 0.30, "Bad": 0.15, "Very Bad": 0.10},
        "election_winner_prediction": {"Democrat": 0.20, "Republican": 0.70, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.20, "State Government": 0.50, "Local Government": 0.30},
        "education_level": {"Less than High School": 0.15, "High School Graduate": 0.30, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.15, "Master's Degree": 0.07, "Doctorate/Professional Degree": 0.03},
        "occupation": {"Healthcare/Medical": 0.15, "Education": 0.10, "Retail/Customer Service": 0.15, "Manufacturing/Production": 0.15, "Construction": 0.10, "Government/Public Service": 0.10, "Agriculture/Farming": 0.08, "Services/Hospitality": 0.07, "Retired": 0.05, "Other": 0.05},
        "monthly_household_income": {"Less than $2,500": 0.20, "$2,500 - $5,000": 0.35, "$5,001 - $7,500": 0.25, "$7,501 - $10,000": 0.10, "Greater than $10,000": 0.10},
        "trust_in_media": {"High Trust": 0.10, "Moderate Trust": 0.25, "Low Trust": 0.35, "No Trust At All": 0.30},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.30, "National News (Cable/Broadcast)": 0.25, "Online News Websites/Apps": 0.20, "Social Media": 0.15, "Friends/Family/Community": 0.07, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.15, "Current Policies Are Adequate": 0.20, "Stricter Policies": 0.40, "Deportation of Undocumented Immigrants": 0.25},
        "climate_change_priority": {"Very High Priority": 0.10, "High Priority": 0.15, "Moderate Priority": 0.25, "Low Priority": 0.30, "Not a Priority": 0.20},
        "gun_control_support": {"Strongly Support More Controls": 0.10, "Support More Controls": 0.15, "Maintain Current Laws": 0.25, "Loosen Laws": 0.30, "Strongly Support Loosening Laws": 0.20},
        "abortion_rights_opinion": {"Legal in All Cases": 0.10, "Legal with Some Restrictions": 0.20, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.35, "Illegal in All Cases": 0.35},
        "party_affiliation": {"Democrats": 0.30, "Republicans": 0.50, "Independent/Unaffiliated": 0.15, "Other": 0.05} # No party registration required, estimates
    },
    # Alaska
    "Alaska": {
        "biggest_issues": {"Healthcare": 0.15, "Unemployment": 0.20, "Education": 0.15, "Inflation": 0.25, "Housing": 0.10, "Crime": 0.15},
        "state_govt_performance": {"Very Good": 0.10, "Good": 0.30, "Average": 0.35, "Bad": 0.15, "Very Bad": 0.10},
        "federal_govt_performance": {"Very Good": 0.08, "Good": 0.20, "Average": 0.25, "Bad": 0.27, "Very Bad": 0.20},
        "opposition_performance": {"Very Good": 0.12, "Good": 0.30, "Average": 0.30, "Bad": 0.18, "Very Bad": 0.10},
        "election_winner_prediction": {"Democrat": 0.30, "Republican": 0.50, "Unsure/Other": 0.20},
        "best_govt_choice": {"Federal Government": 0.25, "State Government": 0.45, "Local Government": 0.30},
        "education_level": {"Less than High School": 0.10, "High School Graduate": 0.25, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.20, "Master's Degree": 0.10, "Doctorate/Professional Degree": 0.05},
        "occupation": {"Government/Public Service": 0.18, "Natural Resources/Energy": 0.15, "Transportation/Logistics": 0.12, "Healthcare/Medical": 0.10, "Retail/Customer Service": 0.10, "Construction": 0.08, "Services/Hospitality": 0.07, "Fishing/Hunting": 0.05, "Retired": 0.05, "Other": 0.10},
        "monthly_household_income": {"Less than $2,500": 0.10, "$2,500 - $5,000": 0.25, "$5,001 - $7,500": 0.30, "$7,501 - $10,000": 0.20, "Greater than $10,000": 0.15},
        "trust_in_media": {"High Trust": 0.15, "Moderate Trust": 0.30, "Low Trust": 0.30, "No Trust At All": 0.25},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.25, "National News (Cable/Broadcast)": 0.30, "Online News Websites/Apps": 0.20, "Social Media": 0.15, "Friends/Family/Community": 0.07, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.20, "Current Policies Are Adequate": 0.25, "Stricter Policies": 0.35, "Deportation of Undocumented Immigrants": 0.20},
        "climate_change_priority": {"Very High Priority": 0.15, "High Priority": 0.20, "Moderate Priority": 0.25, "Low Priority": 0.20, "Not a Priority": 0.20},
        "gun_control_support": {"Strongly Support More Controls": 0.15, "Support More Controls": 0.20, "Maintain Current Laws": 0.30, "Loosen Laws": 0.20, "Strongly Support Loosening Laws": 0.15},
        "abortion_rights_opinion": {"Legal in All Cases": 0.20, "Legal with Some Restrictions": 0.30, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.30, "Illegal in All Cases": 0.20},
        "party_affiliation": {"Democrats": 0.15, "Republicans": 0.23, "Non-Partisan": 0.62, "Other": 0.00} # Highest non-partisan (62%) per USAFacts
    },
    # Arizona
    "Arizona": {
        "biggest_issues": {"Healthcare": 0.18, "Unemployment": 0.15, "Education": 0.15, "Inflation": 0.22, "Housing": 0.15, "Crime": 0.15},
        "state_govt_performance": {"Very Good": 0.10, "Good": 0.30, "Average": 0.35, "Bad": 0.15, "Very Bad": 0.10},
        "federal_govt_performance": {"Very Good": 0.08, "Good": 0.22, "Average": 0.30, "Bad": 0.20, "Very Bad": 0.20},
        "opposition_performance": {"Very Good": 0.12, "Good": 0.30, "Average": 0.30, "Bad": 0.18, "Very Bad": 0.10},
        "election_winner_prediction": {"Democrat": 0.45, "Republican": 0.45, "Unsure/Other": 0.10}, # Swing state
        "best_govt_choice": {"Federal Government": 0.25, "State Government": 0.40, "Local Government": 0.35},
        "education_level": {"Less than High School": 0.10, "High School Graduate": 0.25, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.20, "Master's Degree": 0.10, "Doctorate/Professional Degree": 0.05},
        "occupation": {"Healthcare/Medical": 0.15, "Retail/Customer Service": 0.15, "Construction": 0.12, "Technology/IT": 0.10, "Education": 0.10, "Services/Hospitality": 0.10, "Retired": 0.10, "Finance/Insurance": 0.08, "Other": 0.10},
        "monthly_household_income": {"Less than $2,500": 0.12, "$2,500 - $5,000": 0.28, "$5,001 - $7,500": 0.30, "$7,501 - $10,000": 0.15, "Greater than $10,000": 0.15},
        "trust_in_media": {"High Trust": 0.15, "Moderate Trust": 0.30, "Low Trust": 0.30, "No Trust At All": 0.25},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.25, "National News (Cable/Broadcast)": 0.28, "Online News Websites/Apps": 0.22, "Social Media": 0.15, "Friends/Family/Community": 0.07, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.25, "Current Policies Are Adequate": 0.30, "Stricter Policies": 0.30, "Deportation of Undocumented Immigrants": 0.15},
        "climate_change_priority": {"Very High Priority": 0.15, "High Priority": 0.20, "Moderate Priority": 0.25, "Low Priority": 0.20, "Not a Priority": 0.20},
        "gun_control_support": {"Strongly Support More Controls": 0.20, "Support More Controls": 0.25, "Maintain Current Laws": 0.30, "Loosen Laws": 0.15, "Strongly Support Loosening Laws": 0.10},
        "abortion_rights_opinion": {"Legal in All Cases": 0.25, "Legal with Some Restrictions": 0.35, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.25, "Illegal in All Cases": 0.15},
        "party_affiliation": {"Democrats": 0.31, "Republicans": 0.35, "Independent/Other": 0.34, "Undecided": 0.00} # From USAFacts data (Independent/Other is 34.8%)
    },
    # Arkansas
    "Arkansas": {
        "biggest_issues": {"Healthcare": 0.20, "Unemployment": 0.25, "Education": 0.15, "Inflation": 0.20, "Housing": 0.05, "Crime": 0.15},
        "state_govt_performance": {"Very Good": 0.15, "Good": 0.35, "Average": 0.30, "Bad": 0.10, "Very Bad": 0.10},
        "federal_govt_performance": {"Very Good": 0.05, "Good": 0.15, "Average": 0.20, "Bad": 0.30, "Very Bad": 0.30},
        "opposition_performance": {"Very Good": 0.15, "Good": 0.30, "Average": 0.30, "Bad": 0.15, "Very Bad": 0.10},
        "election_winner_prediction": {"Democrat": 0.25, "Republican": 0.65, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.20, "State Government": 0.50, "Local Government": 0.30},
        "education_level": {"Less than High School": 0.15, "High School Graduate": 0.30, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.15, "Master's Degree": 0.07, "Doctorate/Professional Degree": 0.03},
        "occupation": {"Agriculture/Farming": 0.18, "Manufacturing/Production": 0.15, "Retail/Customer Service": 0.12, "Healthcare/Medical": 0.10, "Education": 0.10, "Services/Hospitality": 0.10, "Government/Public Service": 0.08, "Transportation/Logistics": 0.07, "Retired": 0.05, "Other": 0.05},
        "monthly_household_income": {"Less than $2,500": 0.22, "$2,500 - $5,000": 0.38, "$5,001 - $7,500": 0.20, "$7,501 - $10,000": 0.10, "Greater than $10,000": 0.10},
        "trust_in_media": {"High Trust": 0.08, "Moderate Trust": 0.22, "Low Trust": 0.38, "No Trust At All": 0.32},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.35, "National News (Cable/Broadcast)": 0.20, "Online News Websites/Apps": 0.18, "Social Media": 0.15, "Friends/Family/Community": 0.09, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.10, "Current Policies Are Adequate": 0.20, "Stricter Policies": 0.45, "Deportation of Undocumented Immigrants": 0.25},
        "climate_change_priority": {"Very High Priority": 0.08, "High Priority": 0.12, "Moderate Priority": 0.25, "Low Priority": 0.30, "Not a Priority": 0.25},
        "gun_control_support": {"Strongly Support More Controls": 0.08, "Support More Controls": 0.12, "Maintain Current Laws": 0.28, "Loosen Laws": 0.32, "Strongly Support Loosening Laws": 0.20},
        "abortion_rights_opinion": {"Legal in All Cases": 0.08, "Legal with Some Restrictions": 0.18, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.40, "Illegal in All Cases": 0.34},
        "party_affiliation": {"Democrats": 0.35, "Republicans": 0.60, "Independent/Unaffiliated": 0.05, "Other": 0.00} # No party registration, estimates
    },
    # California
    "California": {
        "biggest_issues": {"Healthcare": 0.15, "Unemployment": 0.10, "Education": 0.12, "Inflation": 0.18, "Housing": 0.30, "Crime": 0.15},
        "state_govt_performance": {"Very Good": 0.15, "Good": 0.40, "Average": 0.30, "Bad": 0.10, "Very Bad": 0.05},
        "federal_govt_performance": {"Very Good": 0.15, "Good": 0.35, "Average": 0.30, "Bad": 0.15, "Very Bad": 0.05},
        "opposition_performance": {"Very Good": 0.08, "Good": 0.25, "Average": 0.35, "Bad": 0.17, "Very Bad": 0.15},
        "election_winner_prediction": {"Democrat": 0.70, "Republican": 0.20, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.40, "State Government": 0.35, "Local Government": 0.25},
        "education_level": {"Less than High School": 0.08, "High School Graduate": 0.20, "Some College": 0.18, "Associate Degree": 0.10, "Bachelor's Degree": 0.25, "Master's Degree": 0.12, "Doctorate/Professional Degree": 0.07},
        "occupation": {"Technology/IT": 0.20, "Healthcare/Medical": 0.12, "Retail/Customer Service": 0.10, "Arts/Entertainment": 0.10, "Education": 0.08, "Services/Hospitality": 0.08, "Finance/Insurance": 0.07, "Construction": 0.05, "Government/Public Service": 0.05, "Retired": 0.05, "Other": 0.10},
        "monthly_household_income": {"Less than $2,500": 0.05, "$2,500 - $5,000": 0.15, "$5,001 - $7,500": 0.25, "$7,501 - $10,000": 0.25, "Greater than $10,000": 0.30},
        "trust_in_media": {"High Trust": 0.25, "Moderate Trust": 0.35, "Low Trust": 0.25, "No Trust At All": 0.15},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.20, "National News (Cable/Broadcast)": 0.30, "Online News Websites/Apps": 0.25, "Social Media": 0.15, "Friends/Family/Community": 0.07, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.40, "Current Policies Are Adequate": 0.30, "Stricter Policies": 0.20, "Deportation of Undocumented Immigrants": 0.10},
        "climate_change_priority": {"Very High Priority": 0.35, "High Priority": 0.30, "Moderate Priority": 0.20, "Low Priority": 0.10, "Not a Priority": 0.05},
        "gun_control_support": {"Strongly Support More Controls": 0.40, "Support More Controls": 0.30, "Maintain Current Laws": 0.15, "Loosen Laws": 0.10, "Strongly Support Loosening Laws": 0.05},
        "abortion_rights_opinion": {"Legal in All Cases": 0.50, "Legal with Some Restrictions": 0.30, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.15, "Illegal in All Cases": 0.05},
        "party_affiliation": {"Democrats": 0.468, "Republicans": 0.239, "No Party Preference": 0.222, "Other": 0.071} # Oct 2023, CA Sec of State
    },
    # Colorado
    "Colorado": {
        "biggest_issues": {"Healthcare": 0.18, "Unemployment": 0.10, "Education": 0.15, "Inflation": 0.20, "Housing": 0.20, "Crime": 0.17},
        "state_govt_performance": {"Very Good": 0.12, "Good": 0.38, "Average": 0.30, "Bad": 0.12, "Very Bad": 0.08},
        "federal_govt_performance": {"Very Good": 0.12, "Good": 0.32, "Average": 0.30, "Bad": 0.18, "Very Bad": 0.08},
        "opposition_performance": {"Very Good": 0.10, "Good": 0.28, "Average": 0.35, "Bad": 0.19, "Very Bad": 0.08},
        "election_winner_prediction": {"Democrat": 0.60, "Republican": 0.30, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.35, "State Government": 0.40, "Local Government": 0.25},
        "education_level": {"Less than High School": 0.07, "High School Graduate": 0.20, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.25, "Master's Degree": 0.10, "Doctorate/Professional Degree": 0.08},
        "occupation": {"Technology/IT": 0.15, "Healthcare/Medical": 0.12, "Retail/Customer Service": 0.10, "Education": 0.10, "Services/Hospitality": 0.10, "Construction": 0.08, "Finance/Insurance": 0.08, "Government/Public Service": 0.07, "Retired": 0.05, "Other": 0.15},
        "monthly_household_income": {"Less than $2,500": 0.07, "$2,500 - $5,000": 0.20, "$5,001 - $7,500": 0.28, "$7,501 - $10,000": 0.25, "Greater than $10,000": 0.20},
        "trust_in_media": {"High Trust": 0.20, "Moderate Trust": 0.35, "Low Trust": 0.28, "No Trust At All": 0.17},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.22, "National News (Cable/Broadcast)": 0.30, "Online News Websites/Apps": 0.25, "Social Media": 0.15, "Friends/Family/Community": 0.05, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.35, "Current Policies Are Adequate": 0.30, "Stricter Policies": 0.25, "Deportation of Undocumented Immigrants": 0.10},
        "climate_change_priority": {"Very High Priority": 0.30, "High Priority": 0.30, "Moderate Priority": 0.20, "Low Priority": 0.10, "Not a Priority": 0.10},
        "gun_control_support": {"Strongly Support More Controls": 0.35, "Support More Controls": 0.30, "Maintain Current Laws": 0.20, "Loosen Laws": 0.10, "Strongly Support Loosening Laws": 0.05},
        "abortion_rights_opinion": {"Legal in All Cases": 0.40, "Legal with Some Restrictions": 0.35, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.15, "Illegal in All Cases": 0.10},
        "party_affiliation": {"Democrats": 0.30, "Republicans": 0.20, "Unaffiliated": 0.49, "Other": 0.01} # High Unaffiliated (49%) per USAFacts
    },
    # Connecticut
    "Connecticut": {
        "biggest_issues": {"Healthcare": 0.18, "Unemployment": 0.10, "Education": 0.15, "Inflation": 0.20, "Housing": 0.20, "Crime": 0.17},
        "state_govt_performance": {"Very Good": 0.12, "Good": 0.38, "Average": 0.30, "Bad": 0.12, "Very Bad": 0.08},
        "federal_govt_performance": {"Very Good": 0.12, "Good": 0.32, "Average": 0.30, "Bad": 0.18, "Very Bad": 0.08},
        "opposition_performance": {"Very Good": 0.10, "Good": 0.28, "Average": 0.35, "Bad": 0.19, "Very Bad": 0.08},
        "election_winner_prediction": {"Democrat": 0.65, "Republican": 0.25, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.35, "State Government": 0.40, "Local Government": 0.25},
        "education_level": {"Less than High School": 0.08, "High School Graduate": 0.20, "Some College": 0.18, "Associate Degree": 0.10, "Bachelor's Degree": 0.25, "Master's Degree": 0.12, "Doctorate/Professional Degree": 0.07},
        "occupation": {"Finance/Insurance": 0.20, "Healthcare/Medical": 0.15, "Education": 0.10, "Retail/Customer Service": 0.10, "Manufacturing/Production": 0.08, "Government/Public Service": 0.07, "Technology/IT": 0.07, "Services/Hospitality": 0.05, "Retired": 0.05, "Other": 0.13},
        "monthly_household_income": {"Less than $2,500": 0.05, "$2,500 - $5,000": 0.15, "$5,001 - $7,500": 0.20, "$7,501 - $10,000": 0.25, "Greater than $10,000": 0.35},
        "trust_in_media": {"High Trust": 0.22, "Moderate Trust": 0.38, "Low Trust": 0.25, "No Trust At All": 0.15},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.20, "National News (Cable/Broadcast)": 0.35, "Online News Websites/Apps": 0.22, "Social Media": 0.10, "Friends/Family/Community": 0.08, "Other": 0.05},
        "immigration_policy_opinion": {"More Lenient Policies": 0.35, "Current Policies Are Adequate": 0.30, "Stricter Policies": 0.25, "Deportation of Undocumented Immigrants": 0.10},
        "climate_change_priority": {"Very High Priority": 0.30, "High Priority": 0.30, "Moderate Priority": 0.20, "Low Priority": 0.10, "Not a Priority": 0.10},
        "gun_control_support": {"Strongly Support More Controls": 0.45, "Support More Controls": 0.30, "Maintain Current Laws": 0.15, "Loosen Laws": 0.07, "Strongly Support Loosening Laws": 0.03},
        "abortion_rights_opinion": {"Legal in All Cases": 0.45, "Legal with Some Restrictions": 0.35, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.15, "Illegal in All Cases": 0.05},
        "party_affiliation": {"Democrats": 0.45, "Republicans": 0.20, "Unaffiliated": 0.30, "Other": 0.05} # Plurality Dem, but significant unaffiliated
    },
    # Delaware
    "Delaware": {
        "biggest_issues": {"Healthcare": 0.18, "Unemployment": 0.10, "Education": 0.15, "Inflation": 0.20, "Housing": 0.20, "Crime": 0.17},
        "state_govt_performance": {"Very Good": 0.12, "Good": 0.38, "Average": 0.30, "Bad": 0.12, "Very Bad": 0.08},
        "federal_govt_performance": {"Very Good": 0.12, "Good": 0.32, "Average": 0.30, "Bad": 0.18, "Very Bad": 0.08},
        "opposition_performance": {"Very Good": 0.10, "Good": 0.28, "Average": 0.35, "Bad": 0.19, "Very Bad": 0.08},
        "election_winner_prediction": {"Democrat": 0.60, "Republican": 0.30, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.35, "State Government": 0.40, "Local Government": 0.25},
        "education_level": {"Less than High School": 0.08, "High School Graduate": 0.22, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.22, "Master's Degree": 0.10, "Doctorate/Professional Degree": 0.08},
        "occupation": {"Finance/Insurance": 0.18, "Healthcare/Medical": 0.15, "Government/Public Service": 0.12, "Education": 0.10, "Retail/Customer Service": 0.10, "Manufacturing/Production": 0.08, "Services/Hospitality": 0.07, "Retired": 0.05, "Other": 0.15},
        "monthly_household_income": {"Less than $2,500": 0.08, "$2,500 - $5,000": 0.20, "$5,001 - $7,500": 0.28, "$7,501 - $10,000": 0.22, "Greater than $10,000": 0.22},
        "trust_in_media": {"High Trust": 0.20, "Moderate Trust": 0.35, "Low Trust": 0.28, "No Trust At All": 0.17},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.22, "National News (Cable/Broadcast)": 0.30, "Online News Websites/Apps": 0.25, "Social Media": 0.15, "Friends/Family/Community": 0.05, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.30, "Current Policies Are Adequate": 0.30, "Stricter Policies": 0.25, "Deportation of Undocumented Immigrants": 0.15},
        "climate_change_priority": {"Very High Priority": 0.28, "High Priority": 0.30, "Moderate Priority": 0.20, "Low Priority": 0.12, "Not a Priority": 0.10},
        "gun_control_support": {"Strongly Support More Controls": 0.40, "Support More Controls": 0.30, "Maintain Current Laws": 0.18, "Loosen Laws": 0.07, "Strongly Support Loosening Laws": 0.05},
        "abortion_rights_opinion": {"Legal in All Cases": 0.40, "Legal with Some Restrictions": 0.35, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.20, "Illegal in All Cases": 0.05},
        "party_affiliation": {"Democrats": 0.45, "Republicans": 0.25, "Unaffiliated": 0.25, "Other": 0.05} # Lean Dem
    },
    # District of Columbia
    "District of Columbia": {
        "biggest_issues": {"Healthcare": 0.15, "Unemployment": 0.10, "Education": 0.10, "Inflation": 0.15, "Housing": 0.25, "Crime": 0.25}, # High housing/crime
        "state_govt_performance": {"Very Good": 0.18, "Good": 0.40, "Average": 0.25, "Bad": 0.10, "Very Bad": 0.07},
        "federal_govt_performance": {"Very Good": 0.20, "Good": 0.40, "Average": 0.25, "Bad": 0.10, "Very Bad": 0.05},
        "opposition_performance": {"Very Good": 0.05, "Good": 0.15, "Average": 0.30, "Bad": 0.25, "Very Bad": 0.25},
        "election_winner_prediction": {"Democrat": 0.85, "Republican": 0.05, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.50, "State Government": 0.25, "Local Government": 0.25}, # More federal focus
        "education_level": {"Less than High School": 0.05, "High School Graduate": 0.15, "Some College": 0.15, "Associate Degree": 0.05, "Bachelor's Degree": 0.30, "Master's Degree": 0.18, "Doctorate/Professional Degree": 0.12}, # Highly educated
        "occupation": {"Government/Public Service": 0.40, "Professional Services": 0.15, "Healthcare/Medical": 0.10, "Education": 0.08, "Retail/Customer Service": 0.07, "Services/Hospitality": 0.05, "Technology/IT": 0.05, "Other": 0.10},
        "monthly_household_income": {"Less than $2,500": 0.03, "$2,500 - $5,000": 0.10, "$5,001 - $7,500": 0.17, "$7,501 - $10,000": 0.25, "Greater than $10,000": 0.45},
        "trust_in_media": {"High Trust": 0.35, "Moderate Trust": 0.40, "Low Trust": 0.15, "No Trust At All": 0.10},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.10, "National News (Cable/Broadcast)": 0.40, "Online News Websites/Apps": 0.30, "Social Media": 0.10, "Friends/Family/Community": 0.07, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.50, "Current Policies Are Adequate": 0.25, "Stricter Policies": 0.15, "Deportation of Undocumented Immigrants": 0.10},
        "climate_change_priority": {"Very High Priority": 0.45, "High Priority": 0.30, "Moderate Priority": 0.15, "Low Priority": 0.05, "Not a Priority": 0.05},
        "gun_control_support": {"Strongly Support More Controls": 0.55, "Support More Controls": 0.30, "Maintain Current Laws": 0.10, "Loosen Laws": 0.03, "Strongly Support Loosening Laws": 0.02},
        "abortion_rights_opinion": {"Legal in All Cases": 0.60, "Legal with Some Restrictions": 0.30, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.07, "Illegal in All Cases": 0.03},
        "party_affiliation": {"Democrats": 0.76, "Republicans": 0.05, "Unaffiliated": 0.15, "Other": 0.04} # Overwhelmingly Dem (76%) per USAFacts
    },
    # Florida
    "Florida": {
        "biggest_issues": {"Healthcare": 0.15, "Unemployment": 0.10, "Education": 0.12, "Inflation": 0.25, "Housing": 0.20, "Crime": 0.18},
        "state_govt_performance": {"Very Good": 0.18, "Good": 0.35, "Average": 0.25, "Bad": 0.12, "Very Bad": 0.10},
        "federal_govt_performance": {"Very Good": 0.08, "Good": 0.22, "Average": 0.25, "Bad": 0.25, "Very Bad": 0.20},
        "opposition_performance": {"Very Good": 0.12, "Good": 0.30, "Average": 0.30, "Bad": 0.18, "Very Bad": 0.10},
        "election_winner_prediction": {"Democrat": 0.45, "Republican": 0.48, "Unsure/Other": 0.07}, # Lean Republican
        "best_govt_choice": {"Federal Government": 0.25, "State Government": 0.40, "Local Government": 0.35},
        "education_level": {"Less than High School": 0.10, "High School Graduate": 0.25, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.20, "Master's Degree": 0.10, "Doctorate/Professional Degree": 0.05},
        "occupation": {"Services/Hospitality": 0.20, "Retail/Customer Service": 0.15, "Healthcare/Medical": 0.12, "Construction": 0.10, "Education": 0.08, "Retired": 0.10, "Finance/Insurance": 0.07, "Government/Public Service": 0.05, "Other": 0.13},
        "monthly_household_income": {"Less than $2,500": 0.10, "$2,500 - $5,000": 0.25, "$5,001 - $7,500": 0.28, "$7,501 - $10,000": 0.17, "Greater than $10,000": 0.20},
        "trust_in_media": {"High Trust": 0.12, "Moderate Trust": 0.28, "Low Trust": 0.30, "No Trust At All": 0.30},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.25, "National News (Cable/Broadcast)": 0.28, "Online News Websites/Apps": 0.22, "Social Media": 0.15, "Friends/Family/Community": 0.07, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.20, "Current Policies Are Adequate": 0.25, "Stricter Policies": 0.35, "Deportation of Undocumented Immigrants": 0.20},
        "climate_change_priority": {"Very High Priority": 0.20, "High Priority": 0.25, "Moderate Priority": 0.25, "Low Priority": 0.15, "Not a Priority": 0.15},
        "gun_control_support": {"Strongly Support More Controls": 0.25, "Support More Controls": 0.25, "Maintain Current Laws": 0.25, "Loosen Laws": 0.15, "Strongly Support Loosening Laws": 0.10},
        "abortion_rights_opinion": {"Legal in All Cases": 0.20, "Legal with Some Restrictions": 0.30, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.30, "Illegal in All Cases": 0.20},
        "party_affiliation": {"Democrats": 0.304, "Republicans": 0.395, "No Party Affiliation": 0.260, "Minor Parties": 0.041} # May 2024, FL Dept of State
    },
    # Georgia
    "Georgia": {
        "biggest_issues": {"Healthcare": 0.18, "Unemployment": 0.15, "Education": 0.15, "Inflation": 0.22, "Housing": 0.15, "Crime": 0.15},
        "state_govt_performance": {"Very Good": 0.10, "Good": 0.30, "Average": 0.35, "Bad": 0.15, "Very Bad": 0.10},
        "federal_govt_performance": {"Very Good": 0.08, "Good": 0.22, "Average": 0.30, "Bad": 0.20, "Very Bad": 0.20},
        "opposition_performance": {"Very Good": 0.12, "Good": 0.30, "Average": 0.30, "Bad": 0.18, "Very Bad": 0.10},
        "election_winner_prediction": {"Democrat": 0.48, "Republican": 0.48, "Unsure/Other": 0.04}, # Swing state
        "best_govt_choice": {"Federal Government": 0.25, "State Government": 0.45, "Local Government": 0.30},
        "education_level": {"Less than High School": 0.12, "High School Graduate": 0.25, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.20, "Master's Degree": 0.10, "Doctorate/Professional Degree": 0.03},
        "occupation": {"Technology/IT": 0.10, "Healthcare/Medical": 0.12, "Retail/Customer Service": 0.15, "Manufacturing/Production": 0.10, "Transportation/Logistics": 0.10, "Education": 0.08, "Government/Public Service": 0.07, "Finance/Insurance": 0.07, "Services/Hospitality": 0.06, "Retired": 0.05, "Other": 0.10},
        "monthly_household_income": {"Less than $2,500": 0.12, "$2,500 - $5,000": 0.28, "$5,001 - $7,500": 0.28, "$7,501 - $10,000": 0.17, "Greater than $10,000": 0.15},
        "trust_in_media": {"High Trust": 0.15, "Moderate Trust": 0.30, "Low Trust": 0.30, "No Trust At All": 0.25},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.25, "National News (Cable/Broadcast)": 0.28, "Online News Websites/Apps": 0.22, "Social Media": 0.15, "Friends/Family/Community": 0.07, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.25, "Current Policies Are Adequate": 0.30, "Stricter Policies": 0.30, "Deportation of Undocumented Immigrants": 0.15},
        "climate_change_priority": {"Very High Priority": 0.20, "High Priority": 0.25, "Moderate Priority": 0.25, "Low Priority": 0.15, "Not a Priority": 0.15},
        "gun_control_support": {"Strongly Support More Controls": 0.25, "Support More Controls": 0.25, "Maintain Current Laws": 0.25, "Loosen Laws": 0.15, "Strongly Support Loosening Laws": 0.10},
        "abortion_rights_opinion": {"Legal in All Cases": 0.25, "Legal with Some Restrictions": 0.35, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.25, "Illegal in All Cases": 0.15},
        "party_affiliation": {"Democrats": 0.40, "Republicans": 0.45, "Independent/Unaffiliated": 0.10, "Other": 0.05} # No party registration required, estimates
    },
    # Hawaii
    "Hawaii": {
        "biggest_issues": {"Healthcare": 0.15, "Unemployment": 0.10, "Education": 0.12, "Inflation": 0.20, "Housing": 0.35, "Crime": 0.08},
        "state_govt_performance": {"Very Good": 0.20, "Good": 0.40, "Average": 0.25, "Bad": 0.10, "Very Bad": 0.05},
        "federal_govt_performance": {"Very Good": 0.18, "Good": 0.38, "Average": 0.25, "Bad": 0.10, "Very Bad": 0.09},
        "opposition_performance": {"Very Good": 0.05, "Good": 0.15, "Average": 0.30, "Bad": 0.25, "Very Bad": 0.25},
        "election_winner_prediction": {"Democrat": 0.75, "Republican": 0.15, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.40, "State Government": 0.35, "Local Government": 0.25},
        "education_level": {"Less than High School": 0.08, "High School Graduate": 0.20, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.22, "Master's Degree": 0.12, "Doctorate/Professional Degree": 0.08},
        "occupation": {"Services/Hospitality": 0.30, "Government/Public Service": 0.15, "Healthcare/Medical": 0.10, "Retail/Customer Service": 0.10, "Education": 0.08, "Military": 0.07, "Construction": 0.05, "Retired": 0.05, "Other": 0.10},
        "monthly_household_income": {"Less than $2,500": 0.08, "$2,500 - $5,000": 0.18, "$5,001 - $7,500": 0.25, "$7,501 - $10,000": 0.22, "Greater than $10,000": 0.27},
        "trust_in_media": {"High Trust": 0.25, "Moderate Trust": 0.40, "Low Trust": 0.20, "No Trust At All": 0.15},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.25, "National News (Cable/Broadcast)": 0.30, "Online News Websites/Apps": 0.25, "Social Media": 0.10, "Friends/Family/Community": 0.07, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.40, "Current Policies Are Adequate": 0.30, "Stricter Policies": 0.20, "Deportation of Undocumented Immigrants": 0.10},
        "climate_change_priority": {"Very High Priority": 0.35, "High Priority": 0.30, "Moderate Priority": 0.20, "Low Priority": 0.10, "Not a Priority": 0.05},
        "gun_control_support": {"Strongly Support More Controls": 0.40, "Support More Controls": 0.30, "Maintain Current Laws": 0.15, "Loosen Laws": 0.10, "Strongly Support Loosening Laws": 0.05},
        "abortion_rights_opinion": {"Legal in All Cases": 0.50, "Legal with Some Restrictions": 0.30, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.15, "Illegal in All Cases": 0.05},
        "party_affiliation": {"Democrats": 0.60, "Republicans": 0.15, "Non-Partisan": 0.20, "Other": 0.05} # Strong Dem, non-partisan option
    },
    # Idaho
    "Idaho": {
        "biggest_issues": {"Healthcare": 0.15, "Unemployment": 0.10, "Education": 0.15, "Inflation": 0.30, "Housing": 0.10, "Crime": 0.10},
        "state_govt_performance": {"Very Good": 0.20, "Good": 0.40, "Average": 0.25, "Bad": 0.10, "Very Bad": 0.05},
        "federal_govt_performance": {"Very Good": 0.03, "Good": 0.10, "Average": 0.15, "Bad": 0.35, "Very Bad": 0.37},
        "opposition_performance": {"Very Good": 0.18, "Good": 0.35, "Average": 0.30, "Bad": 0.10, "Very Bad": 0.07},
        "election_winner_prediction": {"Democrat": 0.15, "Republican": 0.75, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.15, "State Government": 0.50, "Local Government": 0.35},
        "education_level": {"Less than High School": 0.10, "High School Graduate": 0.25, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.20, "Master's Degree": 0.10, "Doctorate/Professional Degree": 0.05},
        "occupation": {"Agriculture/Farming": 0.15, "Forestry/Mining": 0.10, "Manufacturing/Production": 0.10, "Retail/Customer Service": 0.12, "Education": 0.10, "Healthcare/Medical": 0.10, "Construction": 0.10, "Services/Hospitality": 0.08, "Retired": 0.05, "Other": 0.10},
        "monthly_household_income": {"Less than $2,500": 0.10, "$2,500 - $5,000": 0.25, "$5,001 - $7,500": 0.30, "$7,501 - $10,000": 0.20, "Greater than $10,000": 0.15},
        "trust_in_media": {"High Trust": 0.08, "Moderate Trust": 0.20, "Low Trust": 0.35, "No Trust At All": 0.37},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.30, "National News (Cable/Broadcast)": 0.20, "Online News Websites/Apps": 0.20, "Social Media": 0.15, "Friends/Family/Community": 0.10, "Other": 0.05},
        "immigration_policy_opinion": {"More Lenient Policies": 0.10, "Current Policies Are Adequate": 0.15, "Stricter Policies": 0.40, "Deportation of Undocumented Immigrants": 0.35},
        "climate_change_priority": {"Very High Priority": 0.05, "High Priority": 0.10, "Moderate Priority": 0.20, "Low Priority": 0.30, "Not a Priority": 0.35},
        "gun_control_support": {"Strongly Support More Controls": 0.05, "Support More Controls": 0.10, "Maintain Current Laws": 0.20, "Loosen Laws": 0.35, "Strongly Support Loosening Laws": 0.30},
        "abortion_rights_opinion": {"Legal in All Cases": 0.05, "Legal with Some Restrictions": 0.15, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.35, "Illegal in All Cases": 0.45},
        "party_affiliation": {"Democrats": 0.15, "Republicans": 0.70, "Unaffiliated": 0.10, "Other": 0.05} # Very Strong Rep
    },
    # Illinois
    "Illinois": {
        "biggest_issues": {"Healthcare": 0.18, "Unemployment": 0.12, "Education": 0.15, "Inflation": 0.20, "Housing": 0.15, "Crime": 0.20},
        "state_govt_performance": {"Very Good": 0.12, "Good": 0.38, "Average": 0.30, "Bad": 0.12, "Very Bad": 0.08},
        "federal_govt_performance": {"Very Good": 0.12, "Good": 0.32, "Average": 0.30, "Bad": 0.18, "Very Bad": 0.08},
        "opposition_performance": {"Very Good": 0.10, "Good": 0.28, "Average": 0.35, "Bad": 0.19, "Very Bad": 0.08},
        "election_winner_prediction": {"Democrat": 0.60, "Republican": 0.30, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.35, "State Government": 0.40, "Local Government": 0.25},
        "education_level": {"Less than High School": 0.10, "High School Graduate": 0.20, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.25, "Master's Degree": 0.10, "Doctorate/Professional Degree": 0.05},
        "occupation": {"Finance/Insurance": 0.15, "Healthcare/Medical": 0.12, "Education": 0.10, "Retail/Customer Service": 0.12, "Manufacturing/Production": 0.10, "Government/Public Service": 0.08, "Technology/IT": 0.07, "Transportation/Logistics": 0.07, "Services/Hospitality": 0.05, "Retired": 0.05, "Other": 0.09},
        "monthly_household_income": {"Less than $2,500": 0.08, "$2,500 - $5,000": 0.20, "$5,001 - $7,500": 0.28, "$7,501 - $10,000": 0.22, "Greater than $10,000": 0.22},
        "trust_in_media": {"High Trust": 0.20, "Moderate Trust": 0.35, "Low Trust": 0.28, "No Trust At All": 0.17},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.22, "National News (Cable/Broadcast)": 0.30, "Online News Websites/Apps": 0.25, "Social Media": 0.15, "Friends/Family/Community": 0.05, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.30, "Current Policies Are Adequate": 0.30, "Stricter Policies": 0.25, "Deportation of Undocumented Immigrants": 0.15},
        "climate_change_priority": {"Very High Priority": 0.28, "High Priority": 0.30, "Moderate Priority": 0.20, "Low Priority": 0.12, "Not a Priority": 0.10},
        "gun_control_support": {"Strongly Support More Controls": 0.40, "Support More Controls": 0.30, "Maintain Current Laws": 0.18, "Loosen Laws": 0.07, "Strongly Support Loosening Laws": 0.05},
        "abortion_rights_opinion": {"Legal in All Cases": 0.40, "Legal with Some Restrictions": 0.35, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.20, "Illegal in All Cases": 0.05},
        "party_affiliation": {"Democrats": 0.45, "Republicans": 0.30, "Independent/Unaffiliated": 0.20, "Other": 0.05} # No party registration required, estimates
    },
    # Indiana
    "Indiana": {
        "biggest_issues": {"Healthcare": 0.18, "Unemployment": 0.15, "Education": 0.15, "Inflation": 0.22, "Housing": 0.10, "Crime": 0.20},
        "state_govt_performance": {"Very Good": 0.15, "Good": 0.35, "Average": 0.30, "Bad": 0.10, "Very Bad": 0.10},
        "federal_govt_performance": {"Very Good": 0.05, "Good": 0.15, "Average": 0.20, "Bad": 0.30, "Very Bad": 0.30},
        "opposition_performance": {"Very Good": 0.15, "Good": 0.30, "Average": 0.30, "Bad": 0.15, "Very Bad": 0.10},
        "election_winner_prediction": {"Democrat": 0.30, "Republican": 0.60, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.20, "State Government": 0.45, "Local Government": 0.35},
        "education_level": {"Less than High School": 0.12, "High School Graduate": 0.28, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.18, "Master's Degree": 0.08, "Doctorate/Professional Degree": 0.04},
        "occupation": {"Manufacturing/Production": 0.20, "Agriculture/Farming": 0.10, "Retail/Customer Service": 0.12, "Healthcare/Medical": 0.10, "Education": 0.10, "Transportation/Logistics": 0.08, "Services/Hospitality": 0.07, "Construction": 0.05, "Retired": 0.05, "Other": 0.13},
        "monthly_household_income": {"Less than $2,500": 0.15, "$2,500 - $5,000": 0.30, "$5,001 - $7,500": 0.25, "$7,501 - $10,000": 0.15, "Greater than $10,000": 0.15},
        "trust_in_media": {"High Trust": 0.10, "Moderate Trust": 0.25, "Low Trust": 0.35, "No Trust At All": 0.30},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.30, "National News (Cable/Broadcast)": 0.25, "Online News Websites/Apps": 0.20, "Social Media": 0.15, "Friends/Family/Community": 0.07, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.15, "Current Policies Are Adequate": 0.20, "Stricter Policies": 0.40, "Deportation of Undocumented Immigrants": 0.25},
        "climate_change_priority": {"Very High Priority": 0.10, "High Priority": 0.15, "Moderate Priority": 0.25, "Low Priority": 0.30, "Not a Priority": 0.20},
        "gun_control_support": {"Strongly Support More Controls": 0.10, "Support More Controls": 0.15, "Maintain Current Laws": 0.25, "Loosen Laws": 0.30, "Strongly Support Loosening Laws": 0.20},
        "abortion_rights_opinion": {"Legal in All Cases": 0.10, "Legal with Some Restrictions": 0.20, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.35, "Illegal in All Cases": 0.35},
        "party_affiliation": {"Democrats": 0.40, "Republicans": 0.55, "Independent/Unaffiliated": 0.05, "Other": 0.00} # Strong Rep
    },
    # Iowa
    "Iowa": {
        "biggest_issues": {"Healthcare": 0.15, "Unemployment": 0.15, "Education": 0.15, "Inflation": 0.25, "Housing": 0.10, "Crime": 0.20},
        "state_govt_performance": {"Very Good": 0.15, "Good": 0.35, "Average": 0.30, "Bad": 0.10, "Very Bad": 0.10},
        "federal_govt_performance": {"Very Good": 0.08, "Good": 0.20, "Average": 0.25, "Bad": 0.27, "Very Bad": 0.20},
        "opposition_performance": {"Very Good": 0.12, "Good": 0.30, "Average": 0.30, "Bad": 0.18, "Very Bad": 0.10},
        "election_winner_prediction": {"Democrat": 0.40, "Republican": 0.50, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.20, "State Government": 0.50, "Local Government": 0.30},
        "education_level": {"Less than High School": 0.10, "High School Graduate": 0.25, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.20, "Master's Degree": 0.10, "Doctorate/Professional Degree": 0.05},
        "occupation": {"Agriculture/Farming": 0.20, "Manufacturing/Production": 0.15, "Retail/Customer Service": 0.12, "Healthcare/Medical": 0.10, "Education": 0.10, "Services/Hospitality": 0.08, "Finance/Insurance": 0.07, "Government/Public Service": 0.05, "Retired": 0.05, "Other": 0.08},
        "monthly_household_income": {"Less than $2,500": 0.10, "$2,500 - $5,000": 0.25, "$5,001 - $7,500": 0.30, "$7,501 - $10,000": 0.20, "Greater than $10,000": 0.15},
        "trust_in_media": {"High Trust": 0.15, "Moderate Trust": 0.30, "Low Trust": 0.30, "No Trust At All": 0.25},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.30, "National News (Cable/Broadcast)": 0.25, "Online News Websites/Apps": 0.20, "Social Media": 0.15, "Friends/Family/Community": 0.07, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.20, "Current Policies Are Adequate": 0.25, "Stricter Policies": 0.35, "Deportation of Undocumented Immigrants": 0.20},
        "climate_change_priority": {"Very High Priority": 0.15, "High Priority": 0.20, "Moderate Priority": 0.25, "Low Priority": 0.20, "Not a Priority": 0.20},
        "gun_control_support": {"Strongly Support More Controls": 0.20, "Support More Controls": 0.25, "Maintain Current Laws": 0.30, "Loosen Laws": 0.15, "Strongly Support Loosening Laws": 0.10},
        "abortion_rights_opinion": {"Legal in All Cases": 0.20, "Legal with Some Restrictions": 0.30, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.30, "Illegal in All Cases": 0.20},
        "party_affiliation": {"Democrats": 0.30, "Republicans": 0.40, "No Party": 0.25, "Other": 0.05} # Lean Rep
    },
    # Kansas
    "Kansas": {
        "biggest_issues": {"Healthcare": 0.15, "Unemployment": 0.15, "Education": 0.15, "Inflation": 0.25, "Housing": 0.10, "Crime": 0.20},
        "state_govt_performance": {"Very Good": 0.15, "Good": 0.35, "Average": 0.30, "Bad": 0.10, "Very Bad": 0.10},
        "federal_govt_performance": {"Very Good": 0.05, "Good": 0.15, "Average": 0.20, "Bad": 0.30, "Very Bad": 0.30},
        "opposition_performance": {"Very Good": 0.15, "Good": 0.30, "Average": 0.30, "Bad": 0.15, "Very Bad": 0.10},
        "election_winner_prediction": {"Democrat": 0.30, "Republican": 0.60, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.20, "State Government": 0.50, "Local Government": 0.30},
        "education_level": {"Less than High School": 0.10, "High School Graduate": 0.25, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.20, "Master's Degree": 0.10, "Doctorate/Professional Degree": 0.05},
        "occupation": {"Agriculture/Farming": 0.18, "Manufacturing/Production": 0.15, "Retail/Customer Service": 0.12, "Healthcare/Medical": 0.10, "Education": 0.10, "Government/Public Service": 0.08, "Finance/Insurance": 0.07, "Services/Hospitality": 0.05, "Retired": 0.05, "Other": 0.10},
        "monthly_household_income": {"Less than $2,500": 0.10, "$2,500 - $5,000": 0.25, "$5,001 - $7,500": 0.30, "$7,501 - $10,000": 0.20, "Greater than $10,000": 0.15},
        "trust_in_media": {"High Trust": 0.15, "Moderate Trust": 0.30, "Low Trust": 0.30, "No Trust At All": 0.25},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.30, "National News (Cable/Broadcast)": 0.25, "Online News Websites/Apps": 0.20, "Social Media": 0.15, "Friends/Family/Community": 0.07, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.20, "Current Policies Are Adequate": 0.25, "Stricter Policies": 0.35, "Deportation of Undocumented Immigrants": 0.20},
        "climate_change_priority": {"Very High Priority": 0.15, "High Priority": 0.20, "Moderate Priority": 0.25, "Low Priority": 0.20, "Not a Priority": 0.20},
        "gun_control_support": {"Strongly Support More Controls": 0.15, "Support More Controls": 0.20, "Maintain Current Laws": 0.30, "Loosen Laws": 0.20, "Strongly Support Loosening Laws": 0.15},
        "abortion_rights_opinion": {"Legal in All Cases": 0.20, "Legal with Some Restrictions": 0.30, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.30, "Illegal in All Cases": 0.20},
        "party_affiliation": {"Democrats": 0.30, "Republicans": 0.45, "Unaffiliated": 0.20, "Other": 0.05} # Lean Rep, significant unaffiliated
    },
    # Kentucky
    "Kentucky": {
        "biggest_issues": {"Healthcare": 0.25, "Unemployment": 0.20, "Education": 0.15, "Inflation": 0.20, "Housing": 0.05, "Crime": 0.15},
        "state_govt_performance": {"Very Good": 0.15, "Good": 0.35, "Average": 0.30, "Bad": 0.10, "Very Bad": 0.10},
        "federal_govt_performance": {"Very Good": 0.05, "Good": 0.15, "Average": 0.20, "Bad": 0.30, "Very Bad": 0.30},
        "opposition_performance": {"Very Good": 0.15, "Good": 0.30, "Average": 0.30, "Bad": 0.15, "Very Bad": 0.10},
        "election_winner_prediction": {"Democrat": 0.30, "Republican": 0.60, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.20, "State Government": 0.50, "Local Government": 0.30},
        "education_level": {"Less than High School": 0.15, "High School Graduate": 0.30, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.15, "Master's Degree": 0.07, "Doctorate/Professional Degree": 0.03},
        "occupation": {"Manufacturing/Production": 0.18, "Healthcare/Medical": 0.15, "Retail/Customer Service": 0.12, "Education": 0.10, "Agriculture/Farming": 0.10, "Mining": 0.07, "Services/Hospitality": 0.07, "Government/Public Service": 0.05, "Retired": 0.05, "Other": 0.11},
        "monthly_household_income": {"Less than $2,500": 0.18, "$2,500 - $5,000": 0.35, "$5,001 - $7,500": 0.25, "$7,501 - $10,000": 0.10, "Greater than $10,000": 0.12},
        "trust_in_media": {"High Trust": 0.10, "Moderate Trust": 0.25, "Low Trust": 0.35, "No Trust At All": 0.30},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.30, "National News (Cable/Broadcast)": 0.25, "Online News Websites/Apps": 0.20, "Social Media": 0.15, "Friends/Family/Community": 0.07, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.15, "Current Policies Are Adequate": 0.20, "Stricter Policies": 0.40, "Deportation of Undocumented Immigrants": 0.25},
        "climate_change_priority": {"Very High Priority": 0.10, "High Priority": 0.15, "Moderate Priority": 0.25, "Low Priority": 0.30, "Not a Priority": 0.20},
        "gun_control_support": {"Strongly Support More Controls": 0.10, "Support More Controls": 0.15, "Maintain Current Laws": 0.25, "Loosen Laws": 0.30, "Strongly Support Loosening Laws": 0.20},
        "abortion_rights_opinion": {"Legal in All Cases": 0.10, "Legal with Some Restrictions": 0.20, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.35, "Illegal in All Cases": 0.35},
        "party_affiliation": {"Democrats": 0.45, "Republicans": 0.45, "Independent/Other": 0.10, "Undecided": 0.00} # Historically more Dem registration but votes Rep
    },
    # Louisiana
    "Louisiana": {
        "biggest_issues": {"Healthcare": 0.25, "Unemployment": 0.20, "Education": 0.15, "Inflation": 0.20, "Housing": 0.05, "Crime": 0.15},
        "state_govt_performance": {"Very Good": 0.15, "Good": 0.35, "Average": 0.30, "Bad": 0.10, "Very Bad": 0.10},
        "federal_govt_performance": {"Very Good": 0.05, "Good": 0.15, "Average": 0.20, "Bad": 0.30, "Very Bad": 0.30},
        "opposition_performance": {"Very Good": 0.15, "Good": 0.30, "Average": 0.30, "Bad": 0.15, "Very Bad": 0.10},
        "election_winner_prediction": {"Democrat": 0.35, "Republican": 0.55, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.20, "State Government": 0.50, "Local Government": 0.30},
        "education_level": {"Less than High School": 0.15, "High School Graduate": 0.30, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.15, "Master's Degree": 0.07, "Doctorate/Professional Degree": 0.03},
        "occupation": {"Oil/Gas": 0.15, "Healthcare/Medical": 0.15, "Retail/Customer Service": 0.12, "Education": 0.10, "Manufacturing/Production": 0.10, "Services/Hospitality": 0.10, "Government/Public Service": 0.08, "Agriculture/Farming": 0.05, "Retired": 0.05, "Other": 0.10},
        "monthly_household_income": {"Less than $2,500": 0.18, "$2,500 - $5,000": 0.35, "$5,001 - $7,500": 0.25, "$7,501 - $10,000": 0.10, "Greater than $10,000": 0.12},
        "trust_in_media": {"High Trust": 0.10, "Moderate Trust": 0.25, "Low Trust": 0.35, "No Trust At All": 0.30},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.30, "National News (Cable/Broadcast)": 0.25, "Online News Websites/Apps": 0.20, "Social Media": 0.15, "Friends/Family/Community": 0.07, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.15, "Current Policies Are Adequate": 0.20, "Stricter Policies": 0.40, "Deportation of Undocumented Immigrants": 0.25},
        "climate_change_priority": {"Very High Priority": 0.10, "High Priority": 0.15, "Moderate Priority": 0.25, "Low Priority": 0.30, "Not a Priority": 0.20},
        "gun_control_support": {"Strongly Support More Controls": 0.10, "Support More Controls": 0.15, "Maintain Current Laws": 0.25, "Loosen Laws": 0.30, "Strongly Support Loosening Laws": 0.20},
        "abortion_rights_opinion": {"Legal in All Cases": 0.10, "Legal with Some Restrictions": 0.20, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.35, "Illegal in All Cases": 0.35},
        "party_affiliation": {"Democrats": 0.45, "Republicans": 0.35, "No Party": 0.18, "Other": 0.02} # Plurality Dem, but generally votes Rep. (Data from Louisiana SoS suggests plurality Dem registration)
    },
    # Maine
    "Maine": {
        "biggest_issues": {"Healthcare": 0.20, "Unemployment": 0.10, "Education": 0.15, "Inflation": 0.25, "Housing": 0.15, "Crime": 0.15},
        "state_govt_performance": {"Very Good": 0.12, "Good": 0.38, "Average": 0.30, "Bad": 0.12, "Very Bad": 0.08},
        "federal_govt_performance": {"Very Good": 0.12, "Good": 0.32, "Average": 0.30, "Bad": 0.18, "Very Bad": 0.08},
        "opposition_performance": {"Very Good": 0.10, "Good": 0.28, "Average": 0.35, "Bad": 0.19, "Very Bad": 0.08},
        "election_winner_prediction": {"Democrat": 0.50, "Republican": 0.40, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.25, "State Government": 0.45, "Local Government": 0.30},
        "education_level": {"Less than High School": 0.08, "High School Graduate": 0.22, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.22, "Master's Degree": 0.10, "Doctorate/Professional Degree": 0.08},
        "occupation": {"Services/Hospitality": 0.15, "Healthcare/Medical": 0.15, "Education": 0.12, "Agriculture/Farming": 0.10, "Manufacturing/Production": 0.10, "Retail/Customer Service": 0.10, "Forestry/Fishing": 0.08, "Retired": 0.05, "Other": 0.15},
        "monthly_household_income": {"Less than $2,500": 0.08, "$2,500 - $5,000": 0.20, "$5,001 - $7,500": 0.28, "$7,501 - $10,000": 0.22, "Greater than $10,000": 0.22},
        "trust_in_media": {"High Trust": 0.20, "Moderate Trust": 0.35, "Low Trust": 0.28, "No Trust At All": 0.17},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.22, "National News (Cable/Broadcast)": 0.30, "Online News Websites/Apps": 0.25, "Social Media": 0.15, "Friends/Family/Community": 0.05, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.30, "Current Policies Are Adequate": 0.30, "Stricter Policies": 0.25, "Deportation of Undocumented Immigrants": 0.15},
        "climate_change_priority": {"Very High Priority": 0.28, "High Priority": 0.30, "Moderate Priority": 0.20, "Low Priority": 0.12, "Not a Priority": 0.10},
        "gun_control_support": {"Strongly Support More Controls": 0.40, "Support More Controls": 0.30, "Maintain Current Laws": 0.18, "Loosen Laws": 0.07, "Strongly Support Loosening Laws": 0.05},
        "abortion_rights_opinion": {"Legal in All Cases": 0.40, "Legal with Some Restrictions": 0.35, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.20, "Illegal in All Cases": 0.05},
        "party_affiliation": {"Democrats": 0.35, "Republicans": 0.30, "Unenrolled": 0.30, "Other": 0.05} # Significant unenrolled
    },
    # Maryland
    "Maryland": {
        "biggest_issues": {"Healthcare": 0.18, "Unemployment": 0.10, "Education": 0.15, "Inflation": 0.20, "Housing": 0.20, "Crime": 0.17},
        "state_govt_performance": {"Very Good": 0.15, "Good": 0.40, "Average": 0.30, "Bad": 0.10, "Very Bad": 0.05},
        "federal_govt_performance": {"Very Good": 0.15, "Good": 0.35, "Average": 0.30, "Bad": 0.15, "Very Bad": 0.05},
        "opposition_performance": {"Very Good": 0.08, "Good": 0.25, "Average": 0.35, "Bad": 0.17, "Very Bad": 0.15},
        "election_winner_prediction": {"Democrat": 0.65, "Republican": 0.25, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.40, "State Government": 0.35, "Local Government": 0.25},
        "education_level": {"Less than High School": 0.08, "High School Graduate": 0.18, "Some College": 0.18, "Associate Degree": 0.10, "Bachelor's Degree": 0.25, "Master's Degree": 0.15, "Doctorate/Professional Degree": 0.06},
        "occupation": {"Government/Public Service": 0.20, "Healthcare/Medical": 0.15, "Technology/IT": 0.12, "Education": 0.10, "Finance/Insurance": 0.08, "Retail/Customer Service": 0.08, "Professional Services": 0.07, "Construction": 0.05, "Retired": 0.05, "Other": 0.10},
        "monthly_household_income": {"Less than $2,500": 0.05, "$2,500 - $5,000": 0.15, "$5,001 - $7,500": 0.20, "$7,501 - $10,000": 0.25, "Greater than $10,000": 0.35},
        "trust_in_media": {"High Trust": 0.22, "Moderate Trust": 0.38, "Low Trust": 0.25, "No Trust At All": 0.15},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.20, "National News (Cable/Broadcast)": 0.35, "Online News Websites/Apps": 0.22, "Social Media": 0.10, "Friends/Family/Community": 0.08, "Other": 0.05},
        "immigration_policy_opinion": {"More Lenient Policies": 0.35, "Current Policies Are Adequate": 0.30, "Stricter Policies": 0.25, "Deportation of Undocumented Immigrants": 0.10},
        "climate_change_priority": {"Very High Priority": 0.30, "High Priority": 0.30, "Moderate Priority": 0.20, "Low Priority": 0.10, "Not a Priority": 0.10},
        "gun_control_support": {"Strongly Support More Controls": 0.45, "Support More Controls": 0.30, "Maintain Current Laws": 0.15, "Loosen Laws": 0.07, "Strongly Support Loosening Laws": 0.03},
        "abortion_rights_opinion": {"Legal in All Cases": 0.45, "Legal with Some Restrictions": 0.35, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.15, "Illegal in All Cases": 0.05},
        "party_affiliation": {"Democrats": 0.52, "Republicans": 0.26, "Unaffiliated": 0.20, "Other": 0.02} # High Dem % (52.5%) per USAFacts
    },
    # Massachusetts
    "Massachusetts": {
        "biggest_issues": {"Healthcare": 0.15, "Unemployment": 0.10, "Education": 0.12, "Inflation": 0.18, "Housing": 0.30, "Crime": 0.15},
        "state_govt_performance": {"Very Good": 0.18, "Good": 0.40, "Average": 0.25, "Bad": 0.10, "Very Bad": 0.07},
        "federal_govt_performance": {"Very Good": 0.18, "Good": 0.38, "Average": 0.25, "Bad": 0.10, "Very Bad": 0.09},
        "opposition_performance": {"Very Good": 0.05, "Good": 0.15, "Average": 0.30, "Bad": 0.25, "Very Bad": 0.25},
        "election_winner_prediction": {"Democrat": 0.70, "Republican": 0.20, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.40, "State Government": 0.35, "Local Government": 0.25},
        "education_level": {"Less than High School": 0.05, "High School Graduate": 0.15, "Some College": 0.15, "Associate Degree": 0.08, "Bachelor's Degree": 0.28, "Master's Degree": 0.18, "Doctorate/Professional Degree": 0.11}, # Highly educated
        "occupation": {"Healthcare/Medical": 0.20, "Education": 0.15, "Technology/IT": 0.12, "Finance/Insurance": 0.10, "Retail/Customer Service": 0.08, "Professional Services": 0.07, "Government/Public Service": 0.05, "Arts/Entertainment": 0.05, "Retired": 0.05, "Other": 0.13},
        "monthly_household_income": {"Less than $2,500": 0.03, "$2,500 - $5,000": 0.10, "$5,001 - $7,500": 0.17, "$7,501 - $10,000": 0.25, "Greater than $10,000": 0.45},
        "trust_in_media": {"High Trust": 0.35, "Moderate Trust": 0.40, "Low Trust": 0.15, "No Trust At All": 0.10},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.10, "National News (Cable/Broadcast)": 0.40, "Online News Websites/Apps": 0.30, "Social Media": 0.10, "Friends/Family/Community": 0.07, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.50, "Current Policies Are Adequate": 0.25, "Stricter Policies": 0.15, "Deportation of Undocumented Immigrants": 0.10},
        "climate_change_priority": {"Very High Priority": 0.45, "High Priority": 0.30, "Moderate Priority": 0.15, "Low Priority": 0.05, "Not a Priority": 0.05},
        "gun_control_support": {"Strongly Support More Controls": 0.55, "Support More Controls": 0.30, "Maintain Current Laws": 0.10, "Loosen Laws": 0.03, "Strongly Support Loosening Laws": 0.02},
        "abortion_rights_opinion": {"Legal in All Cases": 0.60, "Legal with Some Restrictions": 0.30, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.07, "Illegal in All Cases": 0.03},
        "party_affiliation": {"Democrats": 0.33, "Republicans": 0.09, "Independent/Unaffiliated": 0.58, "Other": 0.00} # Highest Independent % (64%) per USAFacts, adjust for D/R
    },
    # Michigan
    "Michigan": {
        "biggest_issues": {"Healthcare": 0.18, "Unemployment": 0.15, "Education": 0.15, "Inflation": 0.22, "Housing": 0.10, "Crime": 0.20},
        "state_govt_performance": {"Very Good": 0.10, "Good": 0.30, "Average": 0.35, "Bad": 0.15, "Very Bad": 0.10},
        "federal_govt_performance": {"Very Good": 0.08, "Good": 0.22, "Average": 0.30, "Bad": 0.20, "Very Bad": 0.20},
        "opposition_performance": {"Very Good": 0.12, "Good": 0.30, "Average": 0.30, "Bad": 0.18, "Very Bad": 0.10},
        "election_winner_prediction": {"Democrat": 0.50, "Republican": 0.40, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.30, "State Government": 0.40, "Local Government": 0.30},
        "education_level": {"Less than High School": 0.10, "High School Graduate": 0.25, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.20, "Master's Degree": 0.10, "Doctorate/Professional Degree": 0.05},
        "occupation": {"Manufacturing/Production": 0.20, "Healthcare/Medical": 0.12, "Retail/Customer Service": 0.12, "Education": 0.10, "Technology/IT": 0.08, "Transportation/Logistics": 0.07, "Services/Hospitality": 0.07, "Government/Public Service": 0.05, "Retired": 0.05, "Other": 0.14},
        "monthly_household_income": {"Less than $2,500": 0.10, "$2,500 - $5,000": 0.25, "$5,001 - $7,500": 0.28, "$7,501 - $10,000": 0.17, "Greater than $10,000": 0.20},
        "trust_in_media": {"High Trust": 0.15, "Moderate Trust": 0.30, "Low Trust": 0.30, "No Trust At All": 0.25},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.25, "National News (Cable/Broadcast)": 0.28, "Online News Websites/Apps": 0.22, "Social Media": 0.15, "Friends/Family/Community": 0.07, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.20, "Current Policies Are Adequate": 0.25, "Stricter Policies": 0.35, "Deportation of Undocumented Immigrants": 0.20},
        "climate_change_priority": {"Very High Priority": 0.20, "High Priority": 0.25, "Moderate Priority": 0.25, "Low Priority": 0.15, "Not a Priority": 0.15},
        "gun_control_support": {"Strongly Support More Controls": 0.25, "Support More Controls": 0.25, "Maintain Current Laws": 0.25, "Loosen Laws": 0.15, "Strongly Support Loosening Laws": 0.10},
        "abortion_rights_opinion": {"Legal in All Cases": 0.20, "Legal with Some Restrictions": 0.30, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.30, "Illegal in All Cases": 0.20},
        "party_affiliation": {"Democrats": 0.40, "Republicans": 0.35, "Independent/Unaffiliated": 0.20, "Other": 0.05} # No party registration required, estimates
    },
    # Minnesota
    "Minnesota": {
        "biggest_issues": {"Healthcare": 0.18, "Unemployment": 0.10, "Education": 0.15, "Inflation": 0.20, "Housing": 0.15, "Crime": 0.22},
        "state_govt_performance": {"Very Good": 0.12, "Good": 0.38, "Average": 0.30, "Bad": 0.12, "Very Bad": 0.08},
        "federal_govt_performance": {"Very Good": 0.12, "Good": 0.32, "Average": 0.30, "Bad": 0.18, "Very Bad": 0.08},
        "opposition_performance": {"Very Good": 0.10, "Good": 0.28, "Average": 0.35, "Bad": 0.19, "Very Bad": 0.08},
        "election_winner_prediction": {"Democrat": 0.55, "Republican": 0.35, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.35, "State Government": 0.40, "Local Government": 0.25},
        "education_level": {"Less than High School": 0.07, "High School Graduate": 0.20, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.25, "Master's Degree": 0.10, "Doctorate/Professional Degree": 0.08},
        "occupation": {"Healthcare/Medical": 0.18, "Education": 0.12, "Retail/Customer Service": 0.10, "Agriculture/Farming": 0.10, "Manufacturing/Production": 0.10, "Finance/Insurance": 0.08, "Government/Public Service": 0.07, "Technology/IT": 0.06, "Services/Hospitality": 0.05, "Retired": 0.05, "Other": 0.09},
        "monthly_household_income": {"Less than $2,500": 0.07, "$2,500 - $5,000": 0.20, "$5,001 - $7,500": 0.28, "$7,501 - $10,000": 0.25, "Greater than $10,000": 0.20},
        "trust_in_media": {"High Trust": 0.20, "Moderate Trust": 0.35, "Low Trust": 0.28, "No Trust At All": 0.17},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.22, "National News (Cable/Broadcast)": 0.30, "Online News Websites/Apps": 0.25, "Social Media": 0.15, "Friends/Family/Community": 0.05, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.35, "Current Policies Are Adequate": 0.30, "Stricter Policies": 0.25, "Deportation of Undocumented Immigrants": 0.10},
        "climate_change_priority": {"Very High Priority": 0.30, "High Priority": 0.30, "Moderate Priority": 0.20, "Low Priority": 0.10, "Not a Priority": 0.10},
        "gun_control_support": {"Strongly Support More Controls": 0.35, "Support More Controls": 0.30, "Maintain Current Laws": 0.20, "Loosen Laws": 0.10, "Strongly Support Loosening Laws": 0.05},
        "abortion_rights_opinion": {"Legal in All Cases": 0.40, "Legal with Some Restrictions": 0.35, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.15, "Illegal in All Cases": 0.10},
        "party_affiliation": {"Democrats": 0.40, "Republicans": 0.35, "Independent/Unaffiliated": 0.20, "Other": 0.05} # No party registration required, estimates
    },
    # Mississippi
    "Mississippi": {
        "biggest_issues": {"Healthcare": 0.25, "Unemployment": 0.20, "Education": 0.15, "Inflation": 0.20, "Housing": 0.05, "Crime": 0.15},
        "state_govt_performance": {"Very Good": 0.15, "Good": 0.35, "Average": 0.30, "Bad": 0.10, "Very Bad": 0.10},
        "federal_govt_performance": {"Very Good": 0.05, "Good": 0.15, "Average": 0.20, "Bad": 0.30, "Very Bad": 0.30},
        "opposition_performance": {"Very Good": 0.15, "Good": 0.30, "Average": 0.30, "Bad": 0.15, "Very Bad": 0.10},
        "election_winner_prediction": {"Democrat": 0.20, "Republican": 0.70, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.20, "State Government": 0.50, "Local Government": 0.30},
        "education_level": {"Less than High School": 0.18, "High School Graduate": 0.30, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.12, "Master's Degree": 0.07, "Doctorate/Professional Degree": 0.03},
        "occupation": {"Agriculture/Farming": 0.18, "Manufacturing/Production": 0.15, "Retail/Customer Service": 0.12, "Healthcare/Medical": 0.10, "Education": 0.10, "Services/Hospitality": 0.07, "Government/Public Service": 0.08, "Transportation/Logistics": 0.05, "Retired": 0.05, "Other": 0.10},
        "monthly_household_income": {"Less than $2,500": 0.25, "$2,500 - $5,000": 0.40, "$5,001 - $7,500": 0.20, "$7,501 - $10,000": 0.08, "Greater than $10,000": 0.07},
        "trust_in_media": {"High Trust": 0.08, "Moderate Trust": 0.22, "Low Trust": 0.38, "No Trust At All": 0.32},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.35, "National News (Cable/Broadcast)": 0.20, "Online News Websites/Apps": 0.18, "Social Media": 0.15, "Friends/Family/Community": 0.09, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.10, "Current Policies Are Adequate": 0.20, "Stricter Policies": 0.45, "Deportation of Undocumented Immigrants": 0.25},
        "climate_change_priority": {"Very High Priority": 0.08, "High Priority": 0.12, "Moderate Priority": 0.25, "Low Priority": 0.30, "Not a Priority": 0.25},
        "gun_control_support": {"Strongly Support More Controls": 0.08, "Support More Controls": 0.12, "Maintain Current Laws": 0.28, "Loosen Laws": 0.32, "Strongly Support Loosening Laws": 0.20},
        "abortion_rights_opinion": {"Legal in All Cases": 0.08, "Legal with Some Restrictions": 0.18, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.40, "Illegal in All Cases": 0.34},
        "party_affiliation": {"Democrats": 0.35, "Republicans": 0.60, "Independent/Unaffiliated": 0.05, "Other": 0.00} # No party registration, estimates
    },
    # Missouri
    "Missouri": {
        "biggest_issues": {"Healthcare": 0.18, "Unemployment": 0.15, "Education": 0.15, "Inflation": 0.22, "Housing": 0.10, "Crime": 0.20},
        "state_govt_performance": {"Very Good": 0.15, "Good": 0.35, "Average": 0.30, "Bad": 0.10, "Very Bad": 0.10},
        "federal_govt_performance": {"Very Good": 0.05, "Good": 0.15, "Average": 0.20, "Bad": 0.30, "Very Bad": 0.30},
        "opposition_performance": {"Very Good": 0.15, "Good": 0.30, "Average": 0.30, "Bad": 0.15, "Very Bad": 0.10},
        "election_winner_prediction": {"Democrat": 0.40, "Republican": 0.50, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.20, "State Government": 0.45, "Local Government": 0.35},
        "education_level": {"Less than High School": 0.10, "High School Graduate": 0.25, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.20, "Master's Degree": 0.10, "Doctorate/Professional Degree": 0.05},
        "occupation": {"Healthcare/Medical": 0.15, "Retail/Customer Service": 0.12, "Education": 0.10, "Manufacturing/Production": 0.10, "Agriculture/Farming": 0.10, "Transportation/Logistics": 0.08, "Government/Public Service": 0.07, "Services/Hospitality": 0.07, "Finance/Insurance": 0.05, "Retired": 0.05, "Other": 0.11},
        "monthly_household_income": {"Less than $2,500": 0.12, "$2,500 - $5,000": 0.28, "$5,001 - $7,500": 0.28, "$7,501 - $10,000": 0.17, "Greater than $10,000": 0.15},
        "trust_in_media": {"High Trust": 0.15, "Moderate Trust": 0.30, "Low Trust": 0.30, "No Trust At All": 0.25},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.25, "National News (Cable/Broadcast)": 0.28, "Online News Websites/Apps": 0.22, "Social Media": 0.15, "Friends/Family/Community": 0.07, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.20, "Current Policies Are Adequate": 0.25, "Stricter Policies": 0.35, "Deportation of Undocumented Immigrants": 0.20},
        "climate_change_priority": {"Very High Priority": 0.15, "High Priority": 0.20, "Moderate Priority": 0.25, "Low Priority": 0.20, "Not a Priority": 0.20},
        "gun_control_support": {"Strongly Support More Controls": 0.20, "Support More Controls": 0.25, "Maintain Current Laws": 0.30, "Loosen Laws": 0.15, "Strongly Support Loosening Laws": 0.10},
        "abortion_rights_opinion": {"Legal in All Cases": 0.20, "Legal with Some Restrictions": 0.30, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.30, "Illegal in All Cases": 0.20},
        "party_affiliation": {"Democrats": 0.42, "Republicans": 0.53, "Independent/Unaffiliated": 0.05, "Other": 0.00} # Lean Rep
    },
    # Montana
    "Montana": {
        "biggest_issues": {"Healthcare": 0.15, "Unemployment": 0.10, "Education": 0.15, "Inflation": 0.25, "Housing": 0.15, "Crime": 0.20},
        "state_govt_performance": {"Very Good": 0.15, "Good": 0.35, "Average": 0.30, "Bad": 0.10, "Very Bad": 0.10},
        "federal_govt_performance": {"Very Good": 0.05, "Good": 0.15, "Average": 0.20, "Bad": 0.30, "Very Bad": 0.30},
        "opposition_performance": {"Very Good": 0.15, "Good": 0.30, "Average": 0.30, "Bad": 0.15, "Very Bad": 0.10},
        "election_winner_prediction": {"Democrat": 0.35, "Republican": 0.55, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.20, "State Government": 0.50, "Local Government": 0.30},
        "education_level": {"Less than High School": 0.10, "High School Graduate": 0.25, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.20, "Master's Degree": 0.10, "Doctorate/Professional Degree": 0.05},
        "occupation": {"Agriculture/Farming": 0.15, "Mining/Natural Resources": 0.10, "Services/Hospitality": 0.15, "Construction": 0.10, "Retail/Customer Service": 0.12, "Healthcare/Medical": 0.10, "Education": 0.10, "Government/Public Service": 0.05, "Retired": 0.05, "Other": 0.08},
        "monthly_household_income": {"Less than $2,500": 0.10, "$2,500 - $5,000": 0.25, "$5,001 - $7,500": 0.30, "$7,501 - $10,000": 0.20, "Greater than $10,000": 0.15},
        "trust_in_media": {"High Trust": 0.15, "Moderate Trust": 0.30, "Low Trust": 0.30, "No Trust At All": 0.25},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.25, "National News (Cable/Broadcast)": 0.28, "Online News Websites/Apps": 0.22, "Social Media": 0.15, "Friends/Family/Community": 0.07, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.20, "Current Policies Are Adequate": 0.25, "Stricter Policies": 0.35, "Deportation of Undocumented Immigrants": 0.20},
        "climate_change_priority": {"Very High Priority": 0.15, "High Priority": 0.20, "Moderate Priority": 0.25, "Low Priority": 0.20, "Not a Priority": 0.20},
        "gun_control_support": {"Strongly Support More Controls": 0.20, "Support More Controls": 0.25, "Maintain Current Laws": 0.30, "Loosen Laws": 0.15, "Strongly Support Loosening Laws": 0.10},
        "abortion_rights_opinion": {"Legal in All Cases": 0.20, "Legal with Some Restrictions": 0.30, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.30, "Illegal in All Cases": 0.20},
        "party_affiliation": {"Democrats": 0.30, "Republicans": 0.50, "Independent/Unaffiliated": 0.15, "Other": 0.05} # No party registration, estimates
    },
    # Nebraska
    "Nebraska": {
        "biggest_issues": {"Healthcare": 0.15, "Unemployment": 0.10, "Education": 0.15, "Inflation": 0.25, "Housing": 0.10, "Crime": 0.25},
        "state_govt_performance": {"Very Good": 0.18, "Good": 0.38, "Average": 0.25, "Bad": 0.10, "Very Bad": 0.09},
        "federal_govt_performance": {"Very Good": 0.05, "Good": 0.15, "Average": 0.20, "Bad": 0.30, "Very Bad": 0.30},
        "opposition_performance": {"Very Good": 0.15, "Good": 0.30, "Average": 0.30, "Bad": 0.15, "Very Bad": 0.10},
        "election_winner_prediction": {"Democrat": 0.30, "Republican": 0.60, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.20, "State Government": 0.50, "Local Government": 0.30},
        "education_level": {"Less than High School": 0.10, "High School Graduate": 0.25, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.20, "Master's Degree": 0.10, "Doctorate/Professional Degree": 0.05},
        "occupation": {"Agriculture/Farming": 0.20, "Manufacturing/Production": 0.12, "Retail/Customer Service": 0.12, "Finance/Insurance": 0.10, "Healthcare/Medical": 0.10, "Education": 0.10, "Government/Public Service": 0.08, "Services/Hospitality": 0.05, "Retired": 0.05, "Other": 0.08},
        "monthly_household_income": {"Less than $2,500": 0.10, "$2,500 - $5,000": 0.25, "$5,001 - $7,500": 0.30, "$7,501 - $10,000": 0.20, "Greater than $10,000": 0.15},
        "trust_in_media": {"High Trust": 0.15, "Moderate Trust": 0.30, "Low Trust": 0.30, "No Trust At All": 0.25},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.30, "National News (Cable/Broadcast)": 0.25, "Online News Websites/Apps": 0.20, "Social Media": 0.15, "Friends/Family/Community": 0.07, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.20, "Current Policies Are Adequate": 0.25, "Stricter Policies": 0.35, "Deportation of Undocumented Immigrants": 0.20},
        "climate_change_priority": {"Very High Priority": 0.15, "High Priority": 0.20, "Moderate Priority": 0.25, "Low Priority": 0.20, "Not a Priority": 0.20},
        "gun_control_support": {"Strongly Support More Controls": 0.15, "Support More Controls": 0.20, "Maintain Current Laws": 0.30, "Loosen Laws": 0.20, "Strongly Support Loosening Laws": 0.15},
        "abortion_rights_opinion": {"Legal in All Cases": 0.20, "Legal with Some Restrictions": 0.30, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.30, "Illegal in All Cases": 0.20},
        "party_affiliation": {"Democrats": 0.35, "Republicans": 0.60, "Independent/Unaffiliated": 0.05, "Other": 0.00} # Strong Rep
    },
    # Nevada
    "Nevada": {
        "biggest_issues": {"Healthcare": 0.18, "Unemployment": 0.15, "Education": 0.15, "Inflation": 0.22, "Housing": 0.15, "Crime": 0.15},
        "state_govt_performance": {"Very Good": 0.10, "Good": 0.30, "Average": 0.35, "Bad": 0.15, "Very Bad": 0.10},
        "federal_govt_performance": {"Very Good": 0.08, "Good": 0.22, "Average": 0.30, "Bad": 0.20, "Very Bad": 0.20},
        "opposition_performance": {"Very Good": 0.12, "Good": 0.30, "Average": 0.30, "Bad": 0.18, "Very Bad": 0.10},
        "election_winner_prediction": {"Democrat": 0.48, "Republican": 0.47, "Unsure/Other": 0.05}, # Swing state
        "best_govt_choice": {"Federal Government": 0.25, "State Government": 0.40, "Local Government": 0.35},
        "education_level": {"Less than High School": 0.12, "High School Graduate": 0.25, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.18, "Master's Degree": 0.10, "Doctorate/Professional Degree": 0.05},
        "occupation": {"Services/Hospitality": 0.30, "Retail/Customer Service": 0.15, "Construction": 0.10, "Healthcare/Medical": 0.10, "Education": 0.08, "Gaming": 0.07, "Government/Public Service": 0.05, "Retired": 0.05, "Other": 0.10},
        "monthly_household_income": {"Less than $2,500": 0.12, "$2,500 - $5,000": 0.28, "$5,001 - $7,500": 0.30, "$7,501 - $10,000": 0.15, "Greater than $10,000": 0.15},
        "trust_in_media": {"High Trust": 0.15, "Moderate Trust": 0.30, "Low Trust": 0.30, "No Trust At All": 0.25},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.25, "National News (Cable/Broadcast)": 0.28, "Online News Websites/Apps": 0.22, "Social Media": 0.15, "Friends/Family/Community": 0.07, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.25, "Current Policies Are Adequate": 0.30, "Stricter Policies": 0.30, "Deportation of Undocumented Immigrants": 0.15},
        "climate_change_priority": {"Very High Priority": 0.15, "High Priority": 0.20, "Moderate Priority": 0.25, "Low Priority": 0.20, "Not a Priority": 0.20},
        "gun_control_support": {"Strongly Support More Controls": 0.20, "Support More Controls": 0.25, "Maintain Current Laws": 0.30, "Loosen Laws": 0.15, "Strongly Support Loosening Laws": 0.10},
        "abortion_rights_opinion": {"Legal in All Cases": 0.25, "Legal with Some Restrictions": 0.35, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.25, "Illegal in All Cases": 0.15},
        "party_affiliation": {"Democrats": 0.35, "Republicans": 0.30, "Non-Partisan": 0.30, "Other": 0.05} # Significant non-partisan
    },
    # New Hampshire
    "New Hampshire": {
        "biggest_issues": {"Healthcare": 0.15, "Unemployment": 0.10, "Education": 0.15, "Inflation": 0.25, "Housing": 0.15, "Crime": 0.20},
        "state_govt_performance": {"Very Good": 0.10, "Good": 0.30, "Average": 0.35, "Bad": 0.15, "Very Bad": 0.10},
        "federal_govt_performance": {"Very Good": 0.08, "Good": 0.22, "Average": 0.30, "Bad": 0.20, "Very Bad": 0.20},
        "opposition_performance": {"Very Good": 0.12, "Good": 0.30, "Average": 0.30, "Bad": 0.18, "Very Bad": 0.10},
        "election_winner_prediction": {"Democrat": 0.48, "Republican": 0.47, "Unsure/Other": 0.05}, # Swing state
        "best_govt_choice": {"Federal Government": 0.25, "State Government": 0.45, "Local Government": 0.30},
        "education_level": {"Less than High School": 0.07, "High School Graduate": 0.20, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.25, "Master's Degree": 0.10, "Doctorate/Professional Degree": 0.08},
        "occupation": {"Manufacturing/Production": 0.15, "Retail/Customer Service": 0.12, "Healthcare/Medical": 0.10, "Education": 0.10, "Technology/IT": 0.08, "Finance/Insurance": 0.07, "Construction": 0.07, "Services/Hospitality": 0.05, "Retired": 0.05, "Other": 0.21},
        "monthly_household_income": {"Less than $2,500": 0.07, "$2,500 - $5,000": 0.20, "$5,001 - $7,500": 0.28, "$7,501 - $10,000": 0.25, "Greater than $10,000": 0.20},
        "trust_in_media": {"High Trust": 0.20, "Moderate Trust": 0.35, "Low Trust": 0.28, "No Trust At All": 0.17},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.22, "National News (Cable/Broadcast)": 0.30, "Online News Websites/Apps": 0.25, "Social Media": 0.15, "Friends/Family/Community": 0.05, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.25, "Current Policies Are Adequate": 0.30, "Stricter Policies": 0.30, "Deportation of Undocumented Immigrants": 0.15},
        "climate_change_priority": {"Very High Priority": 0.20, "High Priority": 0.25, "Moderate Priority": 0.25, "Low Priority": 0.15, "Not a Priority": 0.15},
        "gun_control_support": {"Strongly Support More Controls": 0.25, "Support More Controls": 0.25, "Maintain Current Laws": 0.25, "Loosen Laws": 0.15, "Strongly Support Loosening Laws": 0.10},
        "abortion_rights_opinion": {"Legal in All Cases": 0.25, "Legal with Some Restrictions": 0.35, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.25, "Illegal in All Cases": 0.15},
        "party_affiliation": {"Democrats": 0.30, "Republicans": 0.30, "Undeclared": 0.38, "Other": 0.02} # High undeclared
    },
    # New Jersey
    "New Jersey": {
        "biggest_issues": {"Healthcare": 0.18, "Unemployment": 0.10, "Education": 0.15, "Inflation": 0.20, "Housing": 0.20, "Crime": 0.17},
        "state_govt_performance": {"Very Good": 0.12, "Good": 0.38, "Average": 0.30, "Bad": 0.12, "Very Bad": 0.08},
        "federal_govt_performance": {"Very Good": 0.12, "Good": 0.32, "Average": 0.30, "Bad": 0.18, "Very Bad": 0.08},
        "opposition_performance": {"Very Good": 0.10, "Good": 0.28, "Average": 0.35, "Bad": 0.19, "Very Bad": 0.08},
        "election_winner_prediction": {"Democrat": 0.60, "Republican": 0.30, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.35, "State Government": 0.40, "Local Government": 0.25},
        "education_level": {"Less than High School": 0.08, "High School Graduate": 0.20, "Some College": 0.18, "Associate Degree": 0.10, "Bachelor's Degree": 0.25, "Master's Degree": 0.12, "Doctorate/Professional Degree": 0.07},
        "occupation": {"Finance/Insurance": 0.20, "Healthcare/Medical": 0.15, "Technology/IT": 0.10, "Retail/Customer Service": 0.10, "Education": 0.10, "Manufacturing/Production": 0.08, "Government/Public Service": 0.07, "Transportation/Logistics": 0.05, "Retired": 0.05, "Other": 0.10},
        "monthly_household_income": {"Less than $2,500": 0.05, "$2,500 - $5,000": 0.15, "$5,001 - $7,500": 0.20, "$7,501 - $10,000": 0.25, "Greater than $10,000": 0.35},
        "trust_in_media": {"High Trust": 0.22, "Moderate Trust": 0.38, "Low Trust": 0.25, "No Trust At All": 0.15},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.20, "National News (Cable/Broadcast)": 0.35, "Online News Websites/Apps": 0.22, "Social Media": 0.10, "Friends/Family/Community": 0.08, "Other": 0.05},
        "immigration_policy_opinion": {"More Lenient Policies": 0.35, "Current Policies Are Adequate": 0.30, "Stricter Policies": 0.25, "Deportation of Undocumented Immigrants": 0.10},
        "climate_change_priority": {"Very High Priority": 0.30, "High Priority": 0.30, "Moderate Priority": 0.20, "Low Priority": 0.10, "Not a Priority": 0.10},
        "gun_control_support": {"Strongly Support More Controls": 0.45, "Support More Controls": 0.30, "Maintain Current Laws": 0.15, "Loosen Laws": 0.07, "Strongly Support Loosening Laws": 0.03},
        "abortion_rights_opinion": {"Legal in All Cases": 0.45, "Legal with Some Restrictions": 0.35, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.15, "Illegal in All Cases": 0.05},
        "party_affiliation": {"Democrats": 0.42, "Republicans": 0.25, "Unaffiliated": 0.30, "Other": 0.03} # Plurality Dem
    },
    # New Mexico
    "New Mexico": {
        "biggest_issues": {"Healthcare": 0.20, "Unemployment": 0.15, "Education": 0.15, "Inflation": 0.20, "Housing": 0.10, "Crime": 0.20},
        "state_govt_performance": {"Very Good": 0.12, "Good": 0.38, "Average": 0.30, "Bad": 0.12, "Very Bad": 0.08},
        "federal_govt_performance": {"Very Good": 0.12, "Good": 0.32, "Average": 0.30, "Bad": 0.18, "Very Bad": 0.08},
        "opposition_performance": {"Very Good": 0.10, "Good": 0.28, "Average": 0.35, "Bad": 0.19, "Very Bad": 0.08},
        "election_winner_prediction": {"Democrat": 0.55, "Republican": 0.35, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.30, "State Government": 0.40, "Local Government": 0.30},
        "education_level": {"Less than High School": 0.15, "High School Graduate": 0.25, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.18, "Master's Degree": 0.08, "Doctorate/Professional Degree": 0.04},
        "occupation": {"Government/Public Service": 0.20, "Healthcare/Medical": 0.12, "Education": 0.10, "Retail/Customer Service": 0.12, "Construction": 0.08, "Agriculture/Farming": 0.07, "Services/Hospitality": 0.07, "Retired": 0.05, "Other": 0.19},
        "monthly_household_income": {"Less than $2,500": 0.15, "$2,500 - $5,000": 0.30, "$5,001 - $7,500": 0.25, "$7,501 - $10,000": 0.15, "Greater than $10,000": 0.15},
        "trust_in_media": {"High Trust": 0.15, "Moderate Trust": 0.30, "Low Trust": 0.30, "No Trust At All": 0.25},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.25, "National News (Cable/Broadcast)": 0.28, "Online News Websites/Apps": 0.22, "Social Media": 0.15, "Friends/Family/Community": 0.07, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.25, "Current Policies Are Adequate": 0.30, "Stricter Policies": 0.30, "Deportation of Undocumented Immigrants": 0.15},
        "climate_change_priority": {"Very High Priority": 0.20, "High Priority": 0.25, "Moderate Priority": 0.25, "Low Priority": 0.15, "Not a Priority": 0.15},
        "gun_control_support": {"Strongly Support More Controls": 0.25, "Support More Controls": 0.25, "Maintain Current Laws": 0.25, "Loosen Laws": 0.15, "Strongly Support Loosening Laws": 0.10},
        "abortion_rights_opinion": {"Legal in All Cases": 0.25, "Legal with Some Restrictions": 0.35, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.25, "Illegal in All Cases": 0.15},
        "party_affiliation": {"Democrats": 0.45, "Republicans": 0.25, "Unaffiliated": 0.25, "Other": 0.05} # Plurality Dem
    },
    # New York
    "New York": {
        "biggest_issues": {"Healthcare": 0.15, "Unemployment": 0.10, "Education": 0.12, "Inflation": 0.18, "Housing": 0.30, "Crime": 0.15},
        "state_govt_performance": {"Very Good": 0.15, "Good": 0.40, "Average": 0.30, "Bad": 0.10, "Very Bad": 0.05},
        "federal_govt_performance": {"Very Good": 0.15, "Good": 0.35, "Average": 0.30, "Bad": 0.15, "Very Bad": 0.05},
        "opposition_performance": {"Very Good": 0.08, "Good": 0.25, "Average": 0.35, "Bad": 0.17, "Very Bad": 0.15},
        "election_winner_prediction": {"Democrat": 0.70, "Republican": 0.20, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.40, "State Government": 0.35, "Local Government": 0.25},
        "education_level": {"Less than High School": 0.08, "High School Graduate": 0.20, "Some College": 0.18, "Associate Degree": 0.10, "Bachelor's Degree": 0.25, "Master's Degree": 0.12, "Doctorate/Professional Degree": 0.07},
        "occupation": {"Finance/Insurance": 0.20, "Arts/Entertainment": 0.15, "Healthcare/Medical": 0.12, "Retail/Customer Service": 0.10, "Education": 0.10, "Technology/IT": 0.08, "Professional Services": 0.07, "Government/Public Service": 0.05, "Services/Hospitality": 0.05, "Retired": 0.05, "Other": 0.03},
        "monthly_household_income": {"Less than $2,500": 0.05, "$2,500 - $5,000": 0.15, "$5,001 - $7,500": 0.20, "$7,501 - $10,000": 0.25, "Greater than $10,000": 0.35},
        "trust_in_media": {"High Trust": 0.25, "Moderate Trust": 0.35, "Low Trust": 0.25, "No Trust At All": 0.15},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.20, "National News (Cable/Broadcast)": 0.30, "Online News Websites/Apps": 0.25, "Social Media": 0.15, "Friends/Family/Community": 0.07, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.40, "Current Policies Are Adequate": 0.30, "Stricter Policies": 0.20, "Deportation of Undocumented Immigrants": 0.10},
        "climate_change_priority": {"Very High Priority": 0.35, "High Priority": 0.30, "Moderate Priority": 0.20, "Low Priority": 0.10, "Not a Priority": 0.05},
        "gun_control_support": {"Strongly Support More Controls": 0.40, "Support More Controls": 0.30, "Maintain Current Laws": 0.15, "Loosen Laws": 0.10, "Strongly Support Loosening Laws": 0.05},
        "abortion_rights_opinion": {"Legal in All Cases": 0.50, "Legal with Some Restrictions": 0.30, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.15, "Illegal in All Cases": 0.05},
        "party_affiliation": {"Democrats": 0.47, "Republicans": 0.22, "Unaffiliated": 0.25, "Other": 0.06} # Plurality Dem, based on general info
    },
    # North Carolina
    "North Carolina": {
        "biggest_issues": {"Healthcare": 0.18, "Unemployment": 0.15, "Education": 0.15, "Inflation": 0.22, "Housing": 0.15, "Crime": 0.15},
        "state_govt_performance": {"Very Good": 0.10, "Good": 0.30, "Average": 0.35, "Bad": 0.15, "Very Bad": 0.10},
        "federal_govt_performance": {"Very Good": 0.08, "Good": 0.22, "Average": 0.30, "Bad": 0.20, "Very Bad": 0.20},
        "opposition_performance": {"Very Good": 0.12, "Good": 0.30, "Average": 0.30, "Bad": 0.18, "Very Bad": 0.10},
        "election_winner_prediction": {"Democrat": 0.48, "Republican": 0.48, "Unsure/Other": 0.04}, # Swing state
        "best_govt_choice": {"Federal Government": 0.25, "State Government": 0.45, "Local Government": 0.30},
        "education_level": {"Less than High School": 0.10, "High School Graduate": 0.25, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.20, "Master's Degree": 0.10, "Doctorate/Professional Degree": 0.05},
        "occupation": {"Manufacturing/Production": 0.15, "Healthcare/Medical": 0.12, "Education": 0.10, "Retail/Customer Service": 0.12, "Finance/Insurance": 0.10, "Government/Public Service": 0.08, "Technology/IT": 0.07, "Services/Hospitality": 0.07, "Agriculture/Farming": 0.05, "Retired": 0.05, "Other": 0.09},
        "monthly_household_income": {"Less than $2,500": 0.12, "$2,500 - $5,000": 0.28, "$5,001 - $7,500": 0.28, "$7,501 - $10,000": 0.17, "Greater than $10,000": 0.15},
        "trust_in_media": {"High Trust": 0.15, "Moderate Trust": 0.30, "Low Trust": 0.30, "No Trust At All": 0.25},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.25, "National News (Cable/Broadcast)": 0.28, "Online News Websites/Apps": 0.22, "Social Media": 0.15, "Friends/Family/Community": 0.07, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.25, "Current Policies Are Adequate": 0.30, "Stricter Policies": 0.30, "Deportation of Undocumented Immigrants": 0.15},
        "climate_change_priority": {"Very High Priority": 0.20, "High Priority": 0.25, "Moderate Priority": 0.25, "Low Priority": 0.15, "Not a Priority": 0.15},
        "gun_control_support": {"Strongly Support More Controls": 0.25, "Support More Controls": 0.25, "Maintain Current Laws": 0.25, "Loosen Laws": 0.15, "Strongly Support Loosening Laws": 0.10},
        "abortion_rights_opinion": {"Legal in All Cases": 0.25, "Legal with Some Restrictions": 0.35, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.25, "Illegal in All Cases": 0.15},
        "party_affiliation": {"Democrats": 0.35, "Republicans": 0.30, "Unaffiliated": 0.35, "Other": 0.00} # Unaffiliated often largest
    },
    # North Dakota
    "North Dakota": {
        "biggest_issues": {"Healthcare": 0.15, "Unemployment": 0.10, "Education": 0.15, "Inflation": 0.30, "Housing": 0.10, "Crime": 0.10},
        "state_govt_performance": {"Very Good": 0.20, "Good": 0.40, "Average": 0.25, "Bad": 0.10, "Very Bad": 0.05},
        "federal_govt_performance": {"Very Good": 0.03, "Good": 0.10, "Average": 0.15, "Bad": 0.35, "Very Bad": 0.37},
        "opposition_performance": {"Very Good": 0.18, "Good": 0.35, "Average": 0.30, "Bad": 0.10, "Very Bad": 0.07},
        "election_winner_prediction": {"Democrat": 0.15, "Republican": 0.75, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.15, "State Government": 0.50, "Local Government": 0.35},
        "education_level": {"Less than High School": 0.10, "High School Graduate": 0.25, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.20, "Master's Degree": 0.10, "Doctorate/Professional Degree": 0.05},
        "occupation": {"Agriculture/Farming": 0.25, "Oil/Gas": 0.15, "Retail/Customer Service": 0.10, "Healthcare/Medical": 0.10, "Education": 0.10, "Construction": 0.08, "Transportation/Logistics": 0.07, "Government/Public Service": 0.05, "Retired": 0.05, "Other": 0.05},
        "monthly_household_income": {"Less than $2,500": 0.10, "$2,500 - $5,000": 0.25, "$5,001 - $7,500": 0.30, "$7,501 - $10,000": 0.20, "Greater than $10,000": 0.15},
        "trust_in_media": {"High Trust": 0.08, "Moderate Trust": 0.20, "Low Trust": 0.35, "No Trust At All": 0.37},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.30, "National News (Cable/Broadcast)": 0.20, "Online News Websites/Apps": 0.20, "Social Media": 0.15, "Friends/Family/Community": 0.10, "Other": 0.05},
        "immigration_policy_opinion": {"More Lenient Policies": 0.10, "Current Policies Are Adequate": 0.15, "Stricter Policies": 0.40, "Deportation of Undocumented Immigrants": 0.35},
        "climate_change_priority": {"Very High Priority": 0.05, "High Priority": 0.10, "Moderate Priority": 0.20, "Low Priority": 0.30, "Not a Priority": 0.35},
        "gun_control_support": {"Strongly Support More Controls": 0.05, "Support More Controls": 0.10, "Maintain Current Laws": 0.20, "Loosen Laws": 0.35, "Strongly Support Loosening Laws": 0.30},
        "abortion_rights_opinion": {"Legal in All Cases": 0.05, "Legal with Some Restrictions": 0.15, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.35, "Illegal in All Cases": 0.45},
        "party_affiliation": {"Democrats": 0.20, "Republicans": 0.60, "Independent/Unaffiliated": 0.15, "Other": 0.05} # No voter registration (so no party data), estimates
    },
    # Ohio
    "Ohio": {
        "biggest_issues": {"Healthcare": 0.18, "Unemployment": 0.15, "Education": 0.15, "Inflation": 0.22, "Housing": 0.10, "Crime": 0.20},
        "state_govt_performance": {"Very Good": 0.15, "Good": 0.35, "Average": 0.30, "Bad": 0.10, "Very Bad": 0.10},
        "federal_govt_performance": {"Very Good": 0.05, "Good": 0.15, "Average": 0.20, "Bad": 0.30, "Very Bad": 0.30},
        "opposition_performance": {"Very Good": 0.15, "Good": 0.30, "Average": 0.30, "Bad": 0.15, "Very Bad": 0.10},
        "election_winner_prediction": {"Democrat": 0.45, "Republican": 0.50, "Unsure/Other": 0.05}, # Lean Republican
        "best_govt_choice": {"Federal Government": 0.20, "State Government": 0.45, "Local Government": 0.35},
        "education_level": {"Less than High School": 0.10, "High School Graduate": 0.25, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.20, "Master's Degree": 0.10, "Doctorate/Professional Degree": 0.05},
        "occupation": {"Manufacturing/Production": 0.20, "Healthcare/Medical": 0.12, "Retail/Customer Service": 0.12, "Education": 0.10, "Construction": 0.08, "Government/Public Service": 0.07, "Services/Hospitality": 0.07, "Transportation/Logistics": 0.05, "Retired": 0.05, "Other": 0.14},
        "monthly_household_income": {"Less than $2,500": 0.12, "$2,500 - $5,000": 0.28, "$5,001 - $7,500": 0.28, "$7,501 - $10,000": 0.17, "Greater than $10,000": 0.15},
        "trust_in_media": {"High Trust": 0.15, "Moderate Trust": 0.30, "Low Trust": 0.30, "No Trust At All": 0.25},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.25, "National News (Cable/Broadcast)": 0.28, "Online News Websites/Apps": 0.22, "Social Media": 0.15, "Friends/Family/Community": 0.07, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.20, "Current Policies Are Adequate": 0.25, "Stricter Policies": 0.35, "Deportation of Undocumented Immigrants": 0.20},
        "climate_change_priority": {"Very High Priority": 0.15, "High Priority": 0.20, "Moderate Priority": 0.25, "Low Priority": 0.20, "Not a Priority": 0.20},
        "gun_control_support": {"Strongly Support More Controls": 0.20, "Support More Controls": 0.25, "Maintain Current Laws": 0.30, "Loosen Laws": 0.15, "Strongly Support Loosening Laws": 0.10},
        "abortion_rights_opinion": {"Legal in All Cases": 0.20, "Legal with Some Restrictions": 0.30, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.30, "Illegal in All Cases": 0.20},
        "party_affiliation": {"Democrats": 0.35, "Republicans": 0.45, "Independent/Unaffiliated": 0.15, "Other": 0.05} # No party registration required, estimates
    },
    # Oklahoma
    "Oklahoma": {
        "biggest_issues": {"Healthcare": 0.20, "Unemployment": 0.20, "Education": 0.15, "Inflation": 0.20, "Housing": 0.05, "Crime": 0.20},
        "state_govt_performance": {"Very Good": 0.15, "Good": 0.35, "Average": 0.30, "Bad": 0.10, "Very Bad": 0.10},
        "federal_govt_performance": {"Very Good": 0.05, "Good": 0.15, "Average": 0.20, "Bad": 0.30, "Very Bad": 0.30},
        "opposition_performance": {"Very Good": 0.15, "Good": 0.30, "Average": 0.30, "Bad": 0.15, "Very Bad": 0.10},
        "election_winner_prediction": {"Democrat": 0.20, "Republican": 0.70, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.20, "State Government": 0.50, "Local Government": 0.30},
        "education_level": {"Less than High School": 0.15, "High School Graduate": 0.30, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.15, "Master's Degree": 0.07, "Doctorate/Professional Degree": 0.03},
        "occupation": {"Oil/Gas": 0.15, "Agriculture/Farming": 0.15, "Healthcare/Medical": 0.12, "Retail/Customer Service": 0.12, "Education": 0.10, "Manufacturing/Production": 0.10, "Government/Public Service": 0.07, "Services/Hospitality": 0.05, "Retired": 0.05, "Other": 0.09},
        "monthly_household_income": {"Less than $2,500": 0.20, "$2,500 - $5,000": 0.35, "$5,001 - $7,500": 0.25, "$7,501 - $10,000": 0.10, "Greater than $10,000": 0.10},
        "trust_in_media": {"High Trust": 0.10, "Moderate Trust": 0.25, "Low Trust": 0.35, "No Trust At All": 0.30},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.30, "National News (Cable/Broadcast)": 0.25, "Online News Websites/Apps": 0.20, "Social Media": 0.15, "Friends/Family/Community": 0.07, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.15, "Current Policies Are Adequate": 0.20, "Stricter Policies": 0.40, "Deportation of Undocumented Immigrants": 0.25},
        "climate_change_priority": {"Very High Priority": 0.10, "High Priority": 0.15, "Moderate Priority": 0.25, "Low Priority": 0.30, "Not a Priority": 0.20},
        "gun_control_support": {"Strongly Support More Controls": 0.10, "Support More Controls": 0.15, "Maintain Current Laws": 0.25, "Loosen Laws": 0.30, "Strongly Support Loosening Laws": 0.20},
        "abortion_rights_opinion": {"Legal in All Cases": 0.10, "Legal with Some Restrictions": 0.20, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.35, "Illegal in All Cases": 0.35},
        "party_affiliation": {"Democrats": 0.20, "Republicans": 0.70, "Independent/Unaffiliated": 0.10, "Other": 0.00} # Strong Rep
    },
    # Oregon
    "Oregon": {
        "biggest_issues": {"Healthcare": 0.18, "Unemployment": 0.10, "Education": 0.15, "Inflation": 0.20, "Housing": 0.20, "Crime": 0.17},
        "state_govt_performance": {"Very Good": 0.12, "Good": 0.38, "Average": 0.30, "Bad": 0.12, "Very Bad": 0.08},
        "federal_govt_performance": {"Very Good": 0.12, "Good": 0.32, "Average": 0.30, "Bad": 0.18, "Very Bad": 0.08},
        "opposition_performance": {"Very Good": 0.10, "Good": 0.28, "Average": 0.35, "Bad": 0.19, "Very Bad": 0.08},
        "election_winner_prediction": {"Democrat": 0.60, "Republican": 0.30, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.35, "State Government": 0.40, "Local Government": 0.25},
        "education_level": {"Less than High School": 0.08, "High School Graduate": 0.20, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.25, "Master's Degree": 0.10, "Doctorate/Professional Degree": 0.07},
        "occupation": {"Technology/IT": 0.12, "Forestry/Agriculture": 0.10, "Healthcare/Medical": 0.12, "Education": 0.10, "Retail/Customer Service": 0.10, "Services/Hospitality": 0.10, "Manufacturing/Production": 0.08, "Government/Public Service": 0.07, "Construction": 0.05, "Retired": 0.05, "Other": 0.11},
        "monthly_household_income": {"Less than $2,500": 0.07, "$2,500 - $5,000": 0.20, "$5,001 - $7,500": 0.28, "$7,501 - $10,000": 0.25, "Greater than $10,000": 0.20},
        "trust_in_media": {"High Trust": 0.20, "Moderate Trust": 0.35, "Low Trust": 0.28, "No Trust At All": 0.17},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.22, "National News (Cable/Broadcast)": 0.30, "Online News Websites/Apps": 0.25, "Social Media": 0.15, "Friends/Family/Community": 0.05, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.35, "Current Policies Are Adequate": 0.30, "Stricter Policies": 0.25, "Deportation of Undocumented Immigrants": 0.10},
        "climate_change_priority": {"Very High Priority": 0.30, "High Priority": 0.30, "Moderate Priority": 0.20, "Low Priority": 0.10, "Not a Priority": 0.10},
        "gun_control_support": {"Strongly Support More Controls": 0.35, "Support More Controls": 0.30, "Maintain Current Laws": 0.20, "Loosen Laws": 0.10, "Strongly Support Loosening Laws": 0.05},
        "abortion_rights_opinion": {"Legal in All Cases": 0.40, "Legal with Some Restrictions": 0.35, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.15, "Illegal in All Cases": 0.10},
        "party_affiliation": {"Democrats": 0.38, "Republicans": 0.25, "Non-Affiliated": 0.30, "Other": 0.07} # Significant non-affiliated
    },
    # Pennsylvania
    "Pennsylvania": {
        "biggest_issues": {"Healthcare": 0.18, "Unemployment": 0.15, "Education": 0.15, "Inflation": 0.22, "Housing": 0.15, "Crime": 0.15},
        "state_govt_performance": {"Very Good": 0.10, "Good": 0.30, "Average": 0.35, "Bad": 0.15, "Very Bad": 0.10},
        "federal_govt_performance": {"Very Good": 0.08, "Good": 0.22, "Average": 0.30, "Bad": 0.20, "Very Bad": 0.20},
        "opposition_performance": {"Very Good": 0.12, "Good": 0.30, "Average": 0.30, "Bad": 0.18, "Very Bad": 0.10},
        "election_winner_prediction": {"Democrat": 0.48, "Republican": 0.48, "Unsure/Other": 0.04}, # Swing state
        "best_govt_choice": {"Federal Government": 0.25, "State Government": 0.40, "Local Government": 0.35},
        "education_level": {"Less than High School": 0.10, "High School Graduate": 0.25, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.20, "Master's Degree": 0.10, "Doctorate/Professional Degree": 0.05},
        "occupation": {"Healthcare/Medical": 0.15, "Manufacturing/Production": 0.15, "Education": 0.10, "Retail/Customer Service": 0.12, "Finance/Insurance": 0.10, "Construction": 0.08, "Government/Public Service": 0.07, "Services/Hospitality": 0.07, "Retired": 0.05, "Other": 0.11},
        "monthly_household_income": {"Less than $2,500": 0.12, "$2,500 - $5,000": 0.28, "$5,001 - $7,500": 0.28, "$7,501 - $10,000": 0.17, "Greater than $10,000": 0.15},
        "trust_in_media": {"High Trust": 0.15, "Moderate Trust": 0.30, "Low Trust": 0.30, "No Trust At All": 0.25},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.25, "National News (Cable/Broadcast)": 0.28, "Online News Websites/Apps": 0.22, "Social Media": 0.15, "Friends/Family/Community": 0.07, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.25, "Current Policies Are Adequate": 0.30, "Stricter Policies": 0.30, "Deportation of Undocumented Immigrants": 0.15},
        "climate_change_priority": {"Very High Priority": 0.20, "High Priority": 0.25, "Moderate Priority": 0.25, "Low Priority": 0.15, "Not a Priority": 0.15},
        "gun_control_support": {"Strongly Support More Controls": 0.25, "Support More Controls": 0.25, "Maintain Current Laws": 0.25, "Loosen Laws": 0.15, "Strongly Support Loosening Laws": 0.10},
        "abortion_rights_opinion": {"Legal in All Cases": 0.25, "Legal with Some Restrictions": 0.35, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.25, "Illegal in All Cases": 0.15},
        "party_affiliation": {"Democrats": 0.45, "Republicans": 0.35, "No Party Affiliation": 0.16, "Other": 0.04} # Registration data varies, estimates based on voting patterns
    },
    # Rhode Island
    "Rhode Island": {
        "biggest_issues": {"Healthcare": 0.18, "Unemployment": 0.10, "Education": 0.15, "Inflation": 0.20, "Housing": 0.20, "Crime": 0.17},
        "state_govt_performance": {"Very Good": 0.12, "Good": 0.38, "Average": 0.30, "Bad": 0.12, "Very Bad": 0.08},
        "federal_govt_performance": {"Very Good": 0.12, "Good": 0.32, "Average": 0.30, "Bad": 0.18, "Very Bad": 0.08},
        "opposition_performance": {"Very Good": 0.10, "Good": 0.28, "Average": 0.35, "Bad": 0.19, "Very Bad": 0.08},
        "election_winner_prediction": {"Democrat": 0.65, "Republican": 0.25, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.35, "State Government": 0.40, "Local Government": 0.25},
        "education_level": {"Less than High School": 0.08, "High School Graduate": 0.20, "Some College": 0.18, "Associate Degree": 0.10, "Bachelor's Degree": 0.25, "Master's Degree": 0.12, "Doctorate/Professional Degree": 0.07},
        "occupation": {"Healthcare/Medical": 0.18, "Education": 0.12, "Retail/Customer Service": 0.10, "Manufacturing/Production": 0.10, "Services/Hospitality": 0.10, "Finance/Insurance": 0.08, "Government/Public Service": 0.07, "Retired": 0.05, "Other": 0.20},
        "monthly_household_income": {"Less than $2,500": 0.05, "$2,500 - $5,000": 0.15, "$5,001 - $7,500": 0.20, "$7,501 - $10,000": 0.25, "Greater than $10,000": 0.35},
        "trust_in_media": {"High Trust": 0.22, "Moderate Trust": 0.38, "Low Trust": 0.25, "No Trust At All": 0.15},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.20, "National News (Cable/Broadcast)": 0.35, "Online News Websites/Apps": 0.22, "Social Media": 0.10, "Friends/Family/Community": 0.08, "Other": 0.05},
        "immigration_policy_opinion": {"More Lenient Policies": 0.35, "Current Policies Are Adequate": 0.30, "Stricter Policies": 0.25, "Deportation of Undocumented Immigrants": 0.10},
        "climate_change_priority": {"Very High Priority": 0.30, "High Priority": 0.30, "Moderate Priority": 0.20, "Low Priority": 0.10, "Not a Priority": 0.10},
        "gun_control_support": {"Strongly Support More Controls": 0.45, "Support More Controls": 0.30, "Maintain Current Laws": 0.15, "Loosen Laws": 0.07, "Strongly Support Loosening Laws": 0.03},
        "abortion_rights_opinion": {"Legal in All Cases": 0.45, "Legal with Some Restrictions": 0.35, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.15, "Illegal in All Cases": 0.05},
        "party_affiliation": {"Democrats": 0.45, "Republicans": 0.15, "Unaffiliated": 0.35, "Other": 0.05} # Strong Dem
    },
    # South Carolina
    "South Carolina": {
        "biggest_issues": {"Healthcare": 0.20, "Unemployment": 0.20, "Education": 0.15, "Inflation": 0.20, "Housing": 0.05, "Crime": 0.20},
        "state_govt_performance": {"Very Good": 0.15, "Good": 0.35, "Average": 0.30, "Bad": 0.10, "Very Bad": 0.10},
        "federal_govt_performance": {"Very Good": 0.05, "Good": 0.15, "Average": 0.20, "Bad": 0.30, "Very Bad": 0.30},
        "opposition_performance": {"Very Good": 0.15, "Good": 0.30, "Average": 0.30, "Bad": 0.15, "Very Bad": 0.10},
        "election_winner_prediction": {"Democrat": 0.30, "Republican": 0.60, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.20, "State Government": 0.50, "Local Government": 0.30},
        "education_level": {"Less than High School": 0.12, "High School Graduate": 0.28, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.18, "Master's Degree": 0.08, "Doctorate/Professional Degree": 0.04},
        "occupation": {"Manufacturing/Production": 0.18, "Services/Hospitality": 0.15, "Retail/Customer Service": 0.12, "Healthcare/Medical": 0.10, "Education": 0.10, "Government/Public Service": 0.08, "Agriculture/Farming": 0.05, "Retired": 0.05, "Other": 0.17},
        "monthly_household_income": {"Less than $2,500": 0.15, "$2,500 - $5,000": 0.30, "$5,001 - $7,500": 0.25, "$7,501 - $10,000": 0.15, "Greater than $10,000": 0.15},
        "trust_in_media": {"High Trust": 0.10, "Moderate Trust": 0.25, "Low Trust": 0.35, "No Trust At All": 0.30},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.30, "National News (Cable/Broadcast)": 0.25, "Online News Websites/Apps": 0.20, "Social Media": 0.15, "Friends/Family/Community": 0.07, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.15, "Current Policies Are Adequate": 0.20, "Stricter Policies": 0.40, "Deportation of Undocumented Immigrants": 0.25},
        "climate_change_priority": {"Very High Priority": 0.10, "High Priority": 0.15, "Moderate Priority": 0.25, "Low Priority": 0.30, "Not a Priority": 0.20},
        "gun_control_support": {"Strongly Support More Controls": 0.10, "Support More Controls": 0.15, "Maintain Current Laws": 0.25, "Loosen Laws": 0.30, "Strongly Support Loosening Laws": 0.20},
        "abortion_rights_opinion": {"Legal in All Cases": 0.10, "Legal with Some Restrictions": 0.20, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.35, "Illegal in All Cases": 0.35},
        "party_affiliation": {"Democrats": 0.30, "Republicans": 0.50, "Independent/Unaffiliated": 0.15, "Other": 0.05} # No party registration required, estimates
    },
    # South Dakota
    "South Dakota": {
        "biggest_issues": {"Healthcare": 0.15, "Unemployment": 0.10, "Education": 0.15, "Inflation": 0.30, "Housing": 0.10, "Crime": 0.10},
        "state_govt_performance": {"Very Good": 0.20, "Good": 0.40, "Average": 0.25, "Bad": 0.10, "Very Bad": 0.05},
        "federal_govt_performance": {"Very Good": 0.03, "Good": 0.10, "Average": 0.15, "Bad": 0.35, "Very Bad": 0.37},
        "opposition_performance": {"Very Good": 0.18, "Good": 0.35, "Average": 0.30, "Bad": 0.10, "Very Bad": 0.07},
        "election_winner_prediction": {"Democrat": 0.15, "Republican": 0.75, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.15, "State Government": 0.50, "Local Government": 0.35},
        "education_level": {"Less than High School": 0.10, "High School Graduate": 0.25, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.20, "Master's Degree": 0.10, "Doctorate/Professional Degree": 0.05},
        "occupation": {"Agriculture/Farming": 0.25, "Healthcare/Medical": 0.12, "Education": 0.10, "Retail/Customer Service": 0.10, "Government/Public Service": 0.10, "Services/Hospitality": 0.08, "Construction": 0.05, "Retired": 0.05, "Other": 0.15},
        "monthly_household_income": {"Less than $2,500": 0.10, "$2,500 - $5,000": 0.25, "$5,001 - $7,500": 0.30, "$7,501 - $10,000": 0.20, "Greater than $10,000": 0.15},
        "trust_in_media": {"High Trust": 0.08, "Moderate Trust": 0.20, "Low Trust": 0.35, "No Trust At All": 0.37},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.30, "National News (Cable/Broadcast)": 0.20, "Online News Websites/Apps": 0.20, "Social Media": 0.15, "Friends/Family/Community": 0.10, "Other": 0.05},
        "immigration_policy_opinion": {"More Lenient Policies": 0.10, "Current Policies Are Adequate": 0.15, "Stricter Policies": 0.40, "Deportation of Undocumented Immigrants": 0.35},
        "climate_change_priority": {"Very High Priority": 0.05, "High Priority": 0.10, "Moderate Priority": 0.20, "Low Priority": 0.30, "Not a Priority": 0.35},
        "gun_control_support": {"Strongly Support More Controls": 0.05, "Support More Controls": 0.10, "Maintain Current Laws": 0.20, "Loosen Laws": 0.35, "Strongly Support Loosening Laws": 0.30},
        "abortion_rights_opinion": {"Legal in All Cases": 0.05, "Legal with Some Restrictions": 0.15, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.35, "Illegal in All Cases": 0.45},
        "party_affiliation": {"Democrats": 0.25, "Republicans": 0.55, "No Party": 0.15, "Other": 0.05} # Strong Rep
    },
    # Tennessee
    "Tennessee": {
        "biggest_issues": {"Healthcare": 0.20, "Unemployment": 0.20, "Education": 0.15, "Inflation": 0.20, "Housing": 0.05, "Crime": 0.20},
        "state_govt_performance": {"Very Good": 0.15, "Good": 0.35, "Average": 0.30, "Bad": 0.10, "Very Bad": 0.10},
        "federal_govt_performance": {"Very Good": 0.05, "Good": 0.15, "Average": 0.20, "Bad": 0.30, "Very Bad": 0.30},
        "opposition_performance": {"Very Good": 0.15, "Good": 0.30, "Average": 0.30, "Bad": 0.15, "Very Bad": 0.10},
        "election_winner_prediction": {"Democrat": 0.25, "Republican": 0.65, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.20, "State Government": 0.50, "Local Government": 0.30},
        "education_level": {"Less than High School": 0.12, "High School Graduate": 0.28, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.18, "Master's Degree": 0.08, "Doctorate/Professional Degree": 0.04},
        "occupation": {"Manufacturing/Production": 0.18, "Healthcare/Medical": 0.15, "Retail/Customer Service": 0.12, "Education": 0.10, "Services/Hospitality": 0.10, "Transportation/Logistics": 0.08, "Government/Public Service": 0.05, "Retired": 0.05, "Other": 0.17},
        "monthly_household_income": {"Less than $2,500": 0.15, "$2,500 - $5,000": 0.30, "$5,001 - $7,500": 0.25, "$7,501 - $10,000": 0.15, "Greater than $10,000": 0.15},
        "trust_in_media": {"High Trust": 0.10, "Moderate Trust": 0.25, "Low Trust": 0.35, "No Trust At All": 0.30},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.30, "National News (Cable/Broadcast)": 0.25, "Online News Websites/Apps": 0.20, "Social Media": 0.15, "Friends/Family/Community": 0.07, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.15, "Current Policies Are Adequate": 0.20, "Stricter Policies": 0.40, "Deportation of Undocumented Immigrants": 0.25},
        "climate_change_priority": {"Very High Priority": 0.10, "High Priority": 0.15, "Moderate Priority": 0.25, "Low Priority": 0.30, "Not a Priority": 0.20},
        "gun_control_support": {"Strongly Support More Controls": 0.10, "Support More Controls": 0.15, "Maintain Current Laws": 0.25, "Loosen Laws": 0.30, "Strongly Support Loosening Laws": 0.20},
        "abortion_rights_opinion": {"Legal in All Cases": 0.10, "Legal with Some Restrictions": 0.20, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.35, "Illegal in All Cases": 0.35},
        "party_affiliation": {"Democrats": 0.35, "Republicans": 0.58, "Independent/Unaffiliated": 0.07, "Other": 0.00} # Strong Rep
    },
    # Texas
    "Texas": {
        "biggest_issues": {"Healthcare": 0.15, "Unemployment": 0.10, "Education": 0.12, "Inflation": 0.25, "Housing": 0.20, "Crime": 0.18},
        "state_govt_performance": {"Very Good": 0.18, "Good": 0.35, "Average": 0.25, "Bad": 0.12, "Very Bad": 0.10},
        "federal_govt_performance": {"Very Good": 0.08, "Good": 0.22, "Average": 0.25, "Bad": 0.25, "Very Bad": 0.20},
        "opposition_performance": {"Very Good": 0.12, "Good": 0.30, "Average": 0.30, "Bad": 0.18, "Very Bad": 0.10},
        "election_winner_prediction": {"Democrat": 0.40, "Republican": 0.50, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.25, "State Government": 0.40, "Local Government": 0.35},
        "education_level": {"Less than High School": 0.15, "High School Graduate": 0.25, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.18, "Master's Degree": 0.08, "Doctorate/Professional Degree": 0.04},
        "occupation": {"Oil/Gas": 0.15, "Healthcare/Medical": 0.12, "Retail/Customer Service": 0.12, "Technology/IT": 0.10, "Manufacturing/Production": 0.10, "Education": 0.08, "Construction": 0.08, "Services/Hospitality": 0.07, "Finance/Insurance": 0.05, "Government/Public Service": 0.05, "Retired": 0.05, "Other": 0.03},
        "monthly_household_income": {"Less than $2,500": 0.12, "$2,500 - $5,000": 0.28, "$5,001 - $7,500": 0.28, "$7,501 - $10,000": 0.17, "Greater than $10,000": 0.15},
        "trust_in_media": {"High Trust": 0.15, "Moderate Trust": 0.30, "Low Trust": 0.30, "No Trust At All": 0.25},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.25, "National News (Cable/Broadcast)": 0.28, "Online News Websites/Apps": 0.22, "Social Media": 0.15, "Friends/Family/Community": 0.07, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.25, "Current Policies Are Adequate": 0.30, "Stricter Policies": 0.30, "Deportation of Undocumented Immigrants": 0.15},
        "climate_change_priority": {"Very High Priority": 0.20, "High Priority": 0.25, "Moderate Priority": 0.25, "Low Priority": 0.15, "Not a Priority": 0.15},
        "gun_control_support": {"Strongly Support More Controls": 0.25, "Support More Controls": 0.25, "Maintain Current Laws": 0.25, "Loosen Laws": 0.15, "Strongly Support Loosening Laws": 0.10},
        "abortion_rights_opinion": {"Legal in All Cases": 0.25, "Legal with Some Restrictions": 0.35, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.25, "Illegal in All Cases": 0.15},
        "party_affiliation": {"Democrats": 0.40, "Republicans": 0.50, "Undecided": 0.10} # Strong Rep
    },
    # Utah
    "Utah": {
        "biggest_issues": {"Healthcare": 0.15, "Unemployment": 0.10, "Education": 0.20, "Inflation": 0.25, "Housing": 0.15, "Crime": 0.15},
        "state_govt_performance": {"Very Good": 0.20, "Good": 0.40, "Average": 0.25, "Bad": 0.10, "Very Bad": 0.05},
        "federal_govt_performance": {"Very Good": 0.03, "Good": 0.10, "Average": 0.15, "Bad": 0.35, "Very Bad": 0.37},
        "opposition_performance": {"Very Good": 0.18, "Good": 0.35, "Average": 0.30, "Bad": 0.10, "Very Bad": 0.07},
        "election_winner_prediction": {"Democrat": 0.15, "Republican": 0.75, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.15, "State Government": 0.50, "Local Government": 0.35},
        "education_level": {"Less than High School": 0.08, "High School Graduate": 0.20, "Some College": 0.22, "Associate Degree": 0.10, "Bachelor's Degree": 0.25, "Master's Degree": 0.10, "Doctorate/Professional Degree": 0.05}, # Higher "Some College" / Degree
        "occupation": {"Technology/IT": 0.15, "Healthcare/Medical": 0.12, "Education": 0.12, "Retail/Customer Service": 0.10, "Construction": 0.10, "Finance/Insurance": 0.08, "Government/Public Service": 0.07, "Services/Hospitality": 0.05, "Retired": 0.05, "Other": 0.16},
        "monthly_household_income": {"Less than $2,500": 0.07, "$2,500 - $5,000": 0.20, "$5,001 - $7,500": 0.28, "$7,501 - $10,000": 0.25, "Greater than $10,000": 0.20},
        "trust_in_media": {"High Trust": 0.10, "Moderate Trust": 0.25, "Low Trust": 0.35, "No Trust At All": 0.30},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.30, "National News (Cable/Broadcast)": 0.25, "Online News Websites/Apps": 0.20, "Social Media": 0.15, "Friends/Family/Community": 0.07, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.15, "Current Policies Are Adequate": 0.20, "Stricter Policies": 0.40, "Deportation of Undocumented Immigrants": 0.25},
        "climate_change_priority": {"Very High Priority": 0.10, "High Priority": 0.15, "Moderate Priority": 0.25, "Low Priority": 0.30, "Not a Priority": 0.20},
        "gun_control_support": {"Strongly Support More Controls": 0.10, "Support More Controls": 0.15, "Maintain Current Laws": 0.25, "Loosen Laws": 0.30, "Strongly Support Loosening Laws": 0.20},
        "abortion_rights_opinion": {"Legal in All Cases": 0.10, "Legal with Some Restrictions": 0.20, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.35, "Illegal in All Cases": 0.35},
        "party_affiliation": {"Democrats": 0.15, "Republicans": 0.60, "Unaffiliated": 0.20, "Other": 0.05} # Strong Rep
    },
    # Vermont
    "Vermont": {
        "biggest_issues": {"Healthcare": 0.20, "Unemployment": 0.10, "Education": 0.15, "Inflation": 0.25, "Housing": 0.15, "Crime": 0.15},
        "state_govt_performance": {"Very Good": 0.15, "Good": 0.40, "Average": 0.30, "Bad": 0.10, "Very Bad": 0.05},
        "federal_govt_performance": {"Very Good": 0.15, "Good": 0.35, "Average": 0.30, "Bad": 0.15, "Very Bad": 0.05},
        "opposition_performance": {"Very Good": 0.08, "Good": 0.25, "Average": 0.35, "Bad": 0.17, "Very Bad": 0.15},
        "election_winner_prediction": {"Democrat": 0.65, "Republican": 0.25, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.35, "State Government": 0.40, "Local Government": 0.25},
        "education_level": {"Less than High School": 0.07, "High School Graduate": 0.20, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.25, "Master's Degree": 0.10, "Doctorate/Professional Degree": 0.08},
        "occupation": {"Education": 0.15, "Healthcare/Medical": 0.15, "Agriculture/Farming": 0.10, "Services/Hospitality": 0.10, "Retail/Customer Service": 0.10, "Government/Public Service": 0.08, "Manufacturing/Production": 0.07, "Arts/Entertainment": 0.05, "Retired": 0.05, "Other": 0.15},
        "monthly_household_income": {"Less than $2,500": 0.07, "$2,500 - $5,000": 0.20, "$5,001 - $7,500": 0.28, "$7,501 - $10,000": 0.25, "Greater than $10,000": 0.20},
        "trust_in_media": {"High Trust": 0.20, "Moderate Trust": 0.35, "Low Trust": 0.28, "No Trust At All": 0.17},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.22, "National News (Cable/Broadcast)": 0.30, "Online News Websites/Apps": 0.25, "Social Media": 0.15, "Friends/Family/Community": 0.05, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.35, "Current Policies Are Adequate": 0.30, "Stricter Policies": 0.25, "Deportation of Undocumented Immigrants": 0.10},
        "climate_change_priority": {"Very High Priority": 0.30, "High Priority": 0.30, "Moderate Priority": 0.20, "Low Priority": 0.10, "Not a Priority": 0.10},
        "gun_control_support": {"Strongly Support More Controls": 0.35, "Support More Controls": 0.30, "Maintain Current Laws": 0.20, "Loosen Laws": 0.10, "Strongly Support Loosening Laws": 0.05},
        "abortion_rights_opinion": {"Legal in All Cases": 0.40, "Legal with Some Restrictions": 0.35, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.15, "Illegal in All Cases": 0.10},
        "party_affiliation": {"Democrats": 0.65, "Republicans": 0.25, "No Party": 0.45, "Other": 0.05} # High No Party
    },
    # Virginia
    "Virginia": {
        "biggest_issues": {"Healthcare": 0.18, "Unemployment": 0.10, "Education": 0.15, "Inflation": 0.20, "Housing": 0.20, "Crime": 0.17},
        "state_govt_performance": {"Very Good": 0.10, "Good": 0.30, "Average": 0.35, "Bad": 0.15, "Very Bad": 0.10},
        "federal_govt_performance": {"Very Good": 0.08, "Good": 0.22, "Average": 0.30, "Bad": 0.20, "Very Bad": 0.20},
        "opposition_performance": {"Very Good": 0.12, "Good": 0.30, "Average": 0.30, "Bad": 0.18, "Very Bad": 0.10},
        "election_winner_prediction": {"Democrat": 0.55, "Republican": 0.35, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.35, "State Government": 0.40, "Local Government": 0.25},
        "education_level": {"Less than High School": 0.08, "High School Graduate": 0.20, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.25, "Master's Degree": 0.12, "Doctorate/Professional Degree": 0.05},
        "occupation": {"Government/Public Service": 0.20, "Technology/IT": 0.15, "Healthcare/Medical": 0.12, "Education": 0.10, "Retail/Customer Service": 0.10, "Professional Services": 0.08, "Military": 0.07, "Services/Hospitality": 0.05, "Retired": 0.05, "Other": 0.08},
        "monthly_household_income": {"Less than $2,500": 0.07, "$2,500 - $5,000": 0.20, "$5,001 - $7,500": 0.28, "$7,501 - $10,000": 0.25, "Greater than $10,000": 0.20},
        "trust_in_media": {"High Trust": 0.20, "Moderate Trust": 0.35, "Low Trust": 0.28, "No Trust At All": 0.17},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.22, "National News (Cable/Broadcast)": 0.30, "Online News Websites/Apps": 0.25, "Social Media": 0.15, "Friends/Family/Community": 0.05, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.35, "Current Policies Are Adequate": 0.30, "Stricter Policies": 0.25, "Deportation of Undocumented Immigrants": 0.10},
        "climate_change_priority": {"Very High Priority": 0.30, "High Priority": 0.30, "Moderate Priority": 0.20, "Low Priority": 0.10, "Not a Priority": 0.10},
        "gun_control_support": {"Strongly Support More Controls": 0.35, "Support More Controls": 0.30, "Maintain Current Laws": 0.20, "Loosen Laws": 0.10, "Strongly Support Loosening Laws": 0.05},
        "abortion_rights_opinion": {"Legal in All Cases": 0.40, "Legal with Some Restrictions": 0.35, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.15, "Illegal in All Cases": 0.10},
        "party_affiliation": {"Democrats": 0.52, "Republicans": 0.43, "Undecided": 0.05} # Lean Dem
    },
    # Washington
    "Washington": {
        "biggest_issues": {"Healthcare": 0.15, "Unemployment": 0.10, "Education": 0.12, "Inflation": 0.18, "Housing": 0.30, "Crime": 0.15},
        "state_govt_performance": {"Very Good": 0.15, "Good": 0.40, "Average": 0.30, "Bad": 0.10, "Very Bad": 0.05},
        "federal_govt_performance": {"Very Good": 0.15, "Good": 0.35, "Average": 0.30, "Bad": 0.15, "Very Bad": 0.05},
        "opposition_performance": {"Very Good": 0.08, "Good": 0.25, "Average": 0.35, "Bad": 0.17, "Very Bad": 0.15},
        "election_winner_prediction": {"Democrat": 0.65, "Republican": 0.25, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.40, "State Government": 0.35, "Local Government": 0.25},
        "education_level": {"Less than High School": 0.07, "High School Graduate": 0.18, "Some College": 0.18, "Associate Degree": 0.10, "Bachelor's Degree": 0.28, "Master's Degree": 0.12, "Doctorate/Professional Degree": 0.07},
        "occupation": {"Technology/IT": 0.20, "Aerospace": 0.10, "Healthcare/Medical": 0.12, "Retail/Customer Service": 0.10, "Education": 0.08, "Services/Hospitality": 0.08, "Construction": 0.07, "Government/Public Service": 0.05, "Retired": 0.05, "Other": 0.15},
        "monthly_household_income": {"Less than $2,500": 0.05, "$2,500 - $5,000": 0.15, "$5,001 - $7,500": 0.25, "$7,501 - $10,000": 0.25, "Greater than $10,000": 0.30},
        "trust_in_media": {"High Trust": 0.25, "Moderate Trust": 0.35, "Low Trust": 0.25, "No Trust At All": 0.15},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.20, "National News (Cable/Broadcast)": 0.30, "Online News Websites/Apps": 0.25, "Social Media": 0.15, "Friends/Family/Community": 0.07, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.40, "Current Policies Are Adequate": 0.30, "Stricter Policies": 0.20, "Deportation of Undocumented Immigrants": 0.10},
        "climate_change_priority": {"Very High Priority": 0.35, "High Priority": 0.30, "Moderate Priority": 0.20, "Low Priority": 0.10, "Not a Priority": 0.05},
        "gun_control_support": {"Strongly Support More Controls": 0.40, "Support More Controls": 0.30, "Maintain Current Laws": 0.15, "Loosen Laws": 0.10, "Strongly Support Loosening Laws": 0.05},
        "abortion_rights_opinion": {"Legal in All Cases": 0.50, "Legal with Some Restrictions": 0.30, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.15, "Illegal in All Cases": 0.05},
        "party_affiliation": {"Democrats": 0.35, "Republicans": 0.25, "Independent/Unaffiliated": 0.35, "Other": 0.05} # No party registration required, estimates
    },
    # West Virginia
    "West Virginia": {
        "biggest_issues": {"Healthcare": 0.25, "Unemployment": 0.20, "Education": 0.15, "Inflation": 0.20, "Housing": 0.05, "Crime": 0.15},
        "state_govt_performance": {"Very Good": 0.15, "Good": 0.35, "Average": 0.30, "Bad": 0.10, "Very Bad": 0.10},
        "federal_govt_performance": {"Very Good": 0.05, "Good": 0.15, "Average": 0.20, "Bad": 0.30, "Very Bad": 0.30},
        "opposition_performance": {"Very Good": 0.15, "Good": 0.30, "Average": 0.30, "Bad": 0.15, "Very Bad": 0.10},
        "election_winner_prediction": {"Democrat": 0.15, "Republican": 0.75, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.20, "State Government": 0.50, "Local Government": 0.30},
        "education_level": {"Less than High School": 0.18, "High School Graduate": 0.30, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.12, "Master's Degree": 0.07, "Doctorate/Professional Degree": 0.03},
        "occupation": {"Mining": 0.15, "Healthcare/Medical": 0.15, "Manufacturing/Production": 0.10, "Retail/Customer Service": 0.12, "Education": 0.10, "Government/Public Service": 0.08, "Services/Hospitality": 0.07, "Retired": 0.05, "Other": 0.18},
        "monthly_household_income": {"Less than $2,500": 0.25, "$2,500 - $5,000": 0.40, "$5,001 - $7,500": 0.20, "$7,501 - $10,000": 0.08, "Greater than $10,000": 0.07},
        "trust_in_media": {"High Trust": 0.08, "Moderate Trust": 0.22, "Low Trust": 0.38, "No Trust At All": 0.32},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.35, "National News (Cable/Broadcast)": 0.20, "Online News Websites/Apps": 0.18, "Social Media": 0.15, "Friends/Family/Community": 0.09, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.10, "Current Policies Are Adequate": 0.20, "Stricter Policies": 0.45, "Deportation of Undocumented Immigrants": 0.25},
        "climate_change_priority": {"Very High Priority": 0.08, "High Priority": 0.12, "Moderate Priority": 0.25, "Low Priority": 0.30, "Not a Priority": 0.25},
        "gun_control_support": {"Strongly Support More Controls": 0.08, "Support More Controls": 0.12, "Maintain Current Laws": 0.28, "Loosen Laws": 0.32, "Strongly Support Loosening Laws": 0.20},
        "abortion_rights_opinion": {"Legal in All Cases": 0.08, "Legal with Some Restrictions": 0.18, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.40, "Illegal in All Cases": 0.34},
        "party_affiliation": {"Democrats": 0.40, "Republicans": 0.50, "Unaffiliated": 0.08, "Other": 0.02} # Historically high Dem registration, but votes Rep
    },
    # Wisconsin
    "Wisconsin": {
        "biggest_issues": {"Healthcare": 0.18, "Unemployment": 0.15, "Education": 0.15, "Inflation": 0.22, "Housing": 0.15, "Crime": 0.15},
        "state_govt_performance": {"Very Good": 0.10, "Good": 0.30, "Average": 0.35, "Bad": 0.15, "Very Bad": 0.10},
        "federal_govt_performance": {"Very Good": 0.08, "Good": 0.22, "Average": 0.30, "Bad": 0.20, "Very Bad": 0.20},
        "opposition_performance": {"Very Good": 0.12, "Good": 0.30, "Average": 0.30, "Bad": 0.18, "Very Bad": 0.10},
        "election_winner_prediction": {"Democrat": 0.48, "Republican": 0.48, "Unsure/Other": 0.04}, # Swing state
        "best_govt_choice": {"Federal Government": 0.25, "State Government": 0.40, "Local Government": 0.35},
        "education_level": {"Less than High School": 0.10, "High School Graduate": 0.25, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.20, "Master's Degree": 0.10, "Doctorate/Professional Degree": 0.05},
        "occupation": {"Manufacturing/Production": 0.20, "Agriculture/Farming": 0.10, "Healthcare/Medical": 0.12, "Retail/Customer Service": 0.12, "Education": 0.10, "Services/Hospitality": 0.08, "Government/Public Service": 0.07, "Retired": 0.05, "Other": 0.16},
        "monthly_household_income": {"Less than $2,500": 0.12, "$2,500 - $5,000": 0.28, "$5,001 - $7,500": 0.28, "$7,501 - $10,000": 0.17, "Greater than $10,000": 0.15},
        "trust_in_media": {"High Trust": 0.15, "Moderate Trust": 0.30, "Low Trust": 0.30, "No Trust At All": 0.25},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.25, "National News (Cable/Broadcast)": 0.28, "Online News Websites/Apps": 0.22, "Social Media": 0.15, "Friends/Family/Community": 0.07, "Other": 0.03},
        "immigration_policy_opinion": {"More Lenient Policies": 0.25, "Current Policies Are Adequate": 0.30, "Stricter Policies": 0.30, "Deportation of Undocumented Immigrants": 0.15},
        "climate_change_priority": {"Very High Priority": 0.20, "High Priority": 0.25, "Moderate Priority": 0.25, "Low Priority": 0.15, "Not a Priority": 0.15},
        "gun_control_support": {"Strongly Support More Controls": 0.25, "Support More Controls": 0.25, "Maintain Current Laws": 0.25, "Loosen Laws": 0.15, "Strongly Support Loosening Laws": 0.10},
        "abortion_rights_opinion": {"Legal in All Cases": 0.25, "Legal with Some Restrictions": 0.35, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.25, "Illegal in All Cases": 0.15},
        "party_affiliation": {"Democrats": 0.48, "Republicans": 0.48, "Undecided": 0.04} # Swing
    },
    "Wyoming": {
        "biggest_issues": {"Healthcare": 0.15, "Unemployment": 0.10, "Education": 0.15, "Inflation": 0.30, "Housing": 0.10, "Crime": 0.10},
        "state_govt_performance": {"Very Good": 0.20, "Good": 0.40, "Average": 0.25, "Bad": 0.10, "Very Bad": 0.05},
        "federal_govt_performance": {"Very Good": 0.03, "Good": 0.10, "Average": 0.15, "Bad": 0.35, "Very Bad": 0.37},
        "opposition_performance": {"Very Good": 0.18, "Good": 0.35, "Average": 0.30, "Bad": 0.10, "Very Bad": 0.07},
        "election_winner_prediction": {"Democrat": 0.10, "Republican": 0.80, "Unsure/Other": 0.10},
        "best_govt_choice": {"Federal Government": 0.15, "State Government": 0.50, "Local Government": 0.35},
        "education_level": {"Less than High School": 0.10, "High School Graduate": 0.25, "Some College": 0.20, "Associate Degree": 0.10, "Bachelor's Degree": 0.20, "Master's Degree": 0.10, "Doctorate/Professional Degree": 0.05},
        "occupation": {"Mining/Oil/Gas": 0.20, "Agriculture/Farming": 0.15, "Services/Hospitality": 0.12, "Retail/Customer Service": 0.10, "Healthcare/Medical": 0.10, "Education": 0.10, "Government/Public Service": 0.08, "Construction": 0.05, "Retired": 0.05, "Other": 0.05},
        "monthly_household_income": {"Less than $2,500": 0.10, "$2,500 - $5,000": 0.25, "$5,001 - $7,500": 0.30, "$7,501 - $10,000": 0.20, "Greater than $10,000": 0.15},
        "trust_in_media": {"High Trust": 0.08, "Moderate Trust": 0.20, "Low Trust": 0.35, "No Trust At All": 0.37},
        "top_media_source": {"Local News (TV/Radio/Newspaper)": 0.30, "National News (Cable/Broadcast)": 0.20, "Online News Websites/Apps": 0.20, "Social Media": 0.15, "Friends/Family/Community": 0.10, "Other": 0.05},
        "immigration_policy_opinion": {"More Lenient Policies": 0.10, "Current Policies Are Adequate": 0.15, "Stricter Policies": 0.40, "Deportation of Undocumented Immigrants": 0.35},
        "climate_change_priority": {"Very High Priority": 0.05, "High Priority": 0.10, "Moderate Priority": 0.20, "Low Priority": 0.30, "Not a Priority": 0.35},
        "gun_control_support": {"Strongly Support More Controls": 0.05, "Support More Controls": 0.10, "Maintain Current Laws": 0.20, "Loosen Laws": 0.35, "Strongly Support Loosening Laws": 0.30},
        "abortion_rights_opinion": {"Legal in All Cases": 0.05, "Legal with Some Restrictions": 0.15, "Illegal with Exceptions (e.g., rape/incest/life of mother)": 0.35, "Illegal in All Cases": 0.45},
        "party_affiliation": {"Democrats": 0.10, "Republicans": 0.70, "Unaffiliated": 0.15, "Other": 0.05} # Strong Rep
    }
}


def generate_entry(state, age, gender, ethnicity, completed=True, force_pref=None):
    """
    Generates a single survey entry, either complete or partial.
    Uses state-specific distributions if available, otherwise falls back to general random choices.
    """
    entry = {
        "call_date": fake.date_time_this_year().strftime("%Y-%m-%d"),
        "agent_id": random.choice(unique_agents),
        "respondent_name": fake.name(),
        "age": age,
        "gender": gender,
        "state": state,
        "survey_status": "Completed" if completed else "Not Completed",
        "ethnicity": ethnicity,
    }

    # Default survey data values (original ones, to be overridden by state-specific distributions)
    # These are the fallback values if no specific distribution is found for a state/field.
    survey_data = {
        "voting_preference": random.choices(["Democrats", "Republicans", "Undecided"], weights=[40, 35, 25])[0],
        "biggest_issues": random.choice(["Healthcare", "Unemployment", "Education", "Inflation", "Housing", "Crime"]),
        "state_govt_performance": random.choice(["Very Good", "Good", "Average", "Bad", "Very Bad"]),
        "federal_govt_performance": random.choice(["Very Good", "Good", "Average", "Bad", "Very Bad"]),
        "opposition_performance": random.choice(["Very Good", "Good", "Average", "Bad", "Very Bad"]),
        "election_winner_prediction": random.choice(["Democrats", "Republicans", "Independent"]),
        "best_govt_choice": random.choice(["Governor", "Senator", "Mayor"]),
        "education_level": random.choice(["No education", "High School", "Graduate", "Postgraduate"]),
        "occupation": random.choice(["Unemployed", "Skilled Labor", "White Collar", "Self-employed", "Student", "Retired"]),
        "monthly_household_income": random.choice(["<2K", "2K-5K", "5K-10K", "10K-20K", "20K+"]),
        "trust_in_media": random.choice(["High", "Moderate", "Low", "None"]),
        "top_media_source": random.choice(["TV", "Online News", "Social Media", "Newspapers", "Podcasts"]),
        "immigration_policy_opinion": random.choice(["Too Strict", "Balanced", "Too Lenient"]),
        "climate_change_priority": random.choice(["Top Priority", "Important", "Not a Priority", "Deny Climate Change"]),
        "gun_control_support": random.choice(["Strongly Support", "Support", "Oppose", "Strongly Oppose"]),
        "abortion_rights_opinion": random.choice(["Pro-Choice", "Pro-Life", "Mixed/Depends"]),
        "party_affiliation": random.choice(["Democrat", "Republican", "Independent", "Libertarian", "Green", "Other"])
    }

    # Override with state-specific distributions if available
    if state in all_state_data_distributions:
        state_dist = all_state_data_distributions[state]
        for field in survey_fields:
            # Check if the current field has a specific distribution for this state
            if field in state_dist:
                choices = list(state_dist[field].keys())
                weights = list(state_dist[field].values())
                # Use random.choices with the provided weights
                survey_data[field] = random.choices(choices, weights=weights, k=1)[0]
    
    # Handle force_pref for voting_preference, which takes precedence over all other logic
    if force_pref:
        survey_data["voting_preference"] = force_pref

    if completed:
        entry.update(survey_data)
    else:
        # Create a partial response with random missing fields
        # Select a random number of fields to answer (between 3 and 6)
        answered_keys = random.sample(survey_fields, k=random.randint(3, 6))
        for k in survey_fields:
            # If the field is in the answered_keys, use its value from survey_data, otherwise set to None
            entry[k] = survey_data[k] if k in answered_keys else None

    return entry

# Step 1: Create full entries with group balancing
complete_entries = []
for state in us_states_population:
    for age in ages:
        for gender in genders:
            for ethnicity in ethnicities:
                # Determine voting preferences needed for balancing, if applicable
                prefs_needed = random.sample(["Democrats", "Republicans", "Undecided"], 2)
                group = []
                # Add entries with forced preferences for balancing
                for pref in prefs_needed:
                    group.append(generate_entry(state, age, gender, ethnicity, completed=True, force_pref=pref))
                
                # Fill the rest of the group with general completed entries
                total_group_size = random.randint(100, 150)
                while len(group) < total_group_size:
                    group.append(generate_entry(state, age, gender, ethnicity, completed=True))
                complete_entries.extend(group)

# Step 2: Add 23% partial records
num_partial = int(len(complete_entries) * 0.23)
partial_entries = []
for _ in range(num_partial):
    state = random.choice(list(us_states_population.keys()))
    age = random.choice(ages)
    gender = random.choice(genders)
    ethnicity = random.choice(ethnicities)
    partial_entries.append(generate_entry(state, age, gender, ethnicity, completed=False))

# Combine and export
final_df = pd.DataFrame(complete_entries + partial_entries)
final_df.to_excel("usa survey data main.xlsx", index=False)

print(f"âœ… Created {len(complete_entries)} full + {len(partial_entries)} partial = {len(final_df)} total records.")
```

## Balance the data 

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.utils import resample

# Load data
df_demo = pd.read_excel("US Demographic Breakdown.xlsx", sheet_name='Sheet1')
df_survey = pd.read_csv("usa statewise survey data.csv")

# Standardize column names
df_demo.columns = df_demo.columns.str.strip()
df_survey.columns = df_survey.columns.str.strip().str.lower()

# Define columns to preserve
trend_columns = [
    'voting_preference', 'survey_status', 'biggest_issues',
    'state_govt_performance', 'federal_govt_performance', 'opposition_performance',
    'election_winner_prediction', 'best_govt_choice', 'monthly_household_income',
    'trust_in_media', 'top_media_source', 'immigration_policy_opinion',
    'climate_change_priority', 'gun_control_support', 'abortion_rights_opinion',
    'party_affiliation'
]
same_columns = [
    'call_date', 'agent_id', 'respondent_name', 'education_level', 'occupation'
]

# Clean up
for col in ['state', 'age', 'gender', 'ethnicity'] + trend_columns:
    if col in df_survey.columns:
        df_survey[col] = df_survey[col].astype(str).str.strip()

for col in ['State', 'Age_Group', 'Gender', 'Ethnicity']:
    df_demo[col] = df_demo[col].astype(str).str.strip()

# Create demographic group keys
df_survey['group'] = df_survey['ethnicity'] + '_' + df_survey['gender'] + '_' + df_survey['age']
df_demo['group'] = df_demo['Ethnicity'] + '_' + df_demo['Gender'] + '_' + df_demo['Age_Group']

# Final dataset container
final = pd.DataFrame()

for state in df_survey['state'].unique():
    print(f"\nProcessing {state}...")

    df_state = df_survey[df_survey['state'] == state].copy()
    df_demo_state = df_demo[df_demo['State'].str.lower() == state.lower()].copy()

    if df_demo_state.empty:
        print(f"[Skipped] No demographics for {state}")
        continue

    total_target = len(df_state)
    balanced_state = pd.DataFrame()

    for _, demo_row in df_demo_state.iterrows():
        group_key = demo_row['group']
        group_pct = float(str(demo_row['Final_Percentage']).replace('%', ''))
        group_target = int(round(total_target * group_pct / 100))

        df_group = df_state[df_state['group'] == group_key]

        if df_group.empty:
            continue

        # If enough samples, keep as-is
        if len(df_group) >= group_target:
            df_final_group = df_group.sample(n=group_target, random_state=42)
        else:
            dfs = []

            for trend_col in trend_columns:
                if trend_col not in df_group.columns:
                    continue

                class_counts = df_group[trend_col].value_counts(normalize=True)

                for cls_val, cls_pct in class_counts.items():
                    cls_target = int(round(group_target * cls_pct))
                    df_cls = df_group[df_group[trend_col] == cls_val]

                    if df_cls.empty:
                        continue

                    df_resampled = resample(df_cls, replace=True, n_samples=cls_target, random_state=42)
                    dfs.append(df_resampled)

            df_final_group = pd.concat(dfs, ignore_index=True).drop_duplicates()

        balanced_state = pd.concat([balanced_state, df_final_group], ignore_index=True)

    print(f"[Done] Final samples for {state}: {len(balanced_state)}")
    final = pd.concat([final, balanced_state], ignore_index=True)

# Save only specified columns
all_needed_columns = ['state', 'age', 'gender', 'ethnicity'] + trend_columns + same_columns
final = final.loc[:, [col for col in all_needed_columns if col in final.columns]]

final.to_excel("Balanced_US_Survey_With_All_Trends.xlsx", index=False)
print("\nâœ… Balancing complete. Output saved to Balanced_US_Survey_With_All_Trends.xlsx.")
```

## Check for trends

```python
import pandas as pd

# Load data
original = pd.read_csv("usa_statewise_survey.csv")
balanced = pd.read_excel("Balanced_US_Survey_UpsampledOnly.xlsx")

# Clean and generate group column
for df in [original, balanced]:
    for col in ['state', 'age', 'gender', 'ethnicity']:
        df[col] = df[col].astype(str).str.strip()
    df['group'] = df['ethnicity'] + '_' + df['gender'] + '_' + df['age']

# Trend columns to check
trend_columns = [
    'voting_preference', 'survey_status', 'biggest_issues',
    'state_govt_performance', 'federal_govt_performance', 'opposition_performance',
    'election_winner_prediction', 'best_govt_choice', 'monthly_household_income',
    'trust_in_media', 'top_media_source', 'immigration_policy_opinion',
    'climate_change_priority', 'gun_control_support', 'abortion_rights_opinion',
    'party_affiliation'
]

# Track changes
tracking = []

for (state, group), df_bal in balanced.groupby(['state', 'group']):
    df_orig = original[(original['state'] == state) &
                       (original['ethnicity'] + '_' + original['gender'] + '_' + original['age'] == group)]

    original_count = len(df_orig)
    balanced_count = len(df_bal)

    for col in trend_columns:
        if col not in df_bal.columns:
            continue

        orig_dist = df_orig[col].value_counts(normalize=True).to_dict() if original_count > 0 else {}
        bal_dist = df_bal[col].value_counts(normalize=True).to_dict()

        tracking.append({
            'state': state,
            'group': group,
            'column': col,
            'original_count': original_count,
            'balanced_count': balanced_count,
            'added_count': max(0, balanced_count - original_count),
            'original_dist': orig_dist,
            'new_dist': bal_dist
        })

# Save results
df_tracking = pd.DataFrame(tracking)
df_tracking.to_excel("Balancing_Tracking_Report_All_Columns.xlsx", index=False)
print("âœ… Tracking report saved to Balancing_Tracking_Report_All_Columns.xlsx")
```
