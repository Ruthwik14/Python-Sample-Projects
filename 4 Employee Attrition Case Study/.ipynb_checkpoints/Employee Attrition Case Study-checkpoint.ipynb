```python
# Define a function to detect outliers using IQR
def find_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return column[(column < lower_bound) | (column > upper_bound)]

# Find outliers in each column
outliers = {}
for column in numerical_columns.columns:
    outliers[column] = find_outliers(df[column])
outliers_dataframe = pd.DataFrame(outliers)
outliers_dataframe.head(15)
```

##  Section 04: KPI/ Metric based questions

```python
#creating 10 different buckets for duration (Deciles)
df['salary_bins'] = pd.qcut(df['Monthly_Income'],14)
df.head()
```

```python
#create empty data frame
KPI = pd.DataFrame()
print(KPI)
```

```python
# calculating no of time  purchase made by each customer
KPI['Average'] = df.groupby("salary_bins").Monthly_Income.mean().round()
KPI.head(14)
```

```python
# calculating no of time  purchase made by each customer
KPI['Minimum_Value'] = df.groupby("salary_bins").Monthly_Income.min()
KPI.head(14)
```

```python
# calculating no of time  purchase made by each customer
KPI['Maximum_Value'] = df.groupby("salary_bins").Monthly_Income.max()
KPI.head(14)
```

```python
# calculating no of time  purchase made by each customer
KPI['No_Of_Emp'] = df.groupby("salary_bins").size()
KPI.head(14)
```

```python
# calculating no of time  purchase made by each customer
KPI['No_of_people_Attrition'] = df.loc[df['Attrition'] == 'Yes'].groupby("salary_bins").size()
KPI.head(14)
```

```python
# calculating no of time  purchase made by each customer
KPI['Attrition rate'] = KPI['No_of_people_Attrition']/KPI['No_Of_Emp'] *100
KPI.head(14)
```

```python
# calculating no of time  purchase made by each customer
KPI['Attrition rate'] = KPI['Attrition rate'].round(2)
KPI.head(14)
```

```python
# this is RFM (Recency Frequency Monetary) table for all customers
KPI = KPI.reset_index()
KPI.head(14)
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='salary_bins', y='Attrition rate', data=KPI)
plt.title('Attrition Rate by Group')
plt.xlabel('Attrition Group')
plt.ylabel('Attrition Rate')
plt.show()
```

```python
# Pie Chart
plt.figure(figsize=(6, 6))
plt.pie(KPI['Attrition rate'], labels=KPI['salary_bins'], autopct='%1.1f%%')
plt.title('Attrition Rate Distribution')
plt.show()
```

# analyzing Attrition rate for a combination of Two variables, which of the below segments

### Employees who got a promotion recently and spent <1 year with their current manager versus employees who have spent >=1 year.

```python
## Encoding the Attrition Column for Modelling
df['Attrition'] = [1 if i == 'Yes' else 0 for i in df['Attrition']]
```

```python
# Create two categories based on Manager_Years
df['Years_Category'] = df['Years_With_Curr_Manager'].apply(lambda x: '<1 year' if x < 1 else '>=1 year')

df['Emp_promot_ rectly'] = df['Years_Since_Last_Promotion'].apply(lambda x: '<2 year' if x < 2 else '>=1 year')

# Create a simple bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='Years_Category', y='Attrition', data=df)

# Customize the plot
plt.title("Employees who got a promotion recently and spent <1 year with their current manager versus employees who have spent >=1 year.")
plt.xlabel("Years with Current Manager")
plt.ylabel("Attrition Rate")

# Show the plot
plt.show()
```

## Employees who got a Promotion, but salary increment was low versus employee who got a higher increment.

```python
# Create two categories based on Manager_Years
df['Emp_Promoted'] = df['Years_Since_Last_Promotion'].apply(lambda x: 'promoted' if x < 1 else 'not Promoted')

df['Emp_promot_ rectly'] = df['Years_Since_Last_Promotion'].apply(lambda x: '<2 year' if x < 2 else '>=1 year')

# Create a simple bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='Years_Category', y='Attrition', data=df)

# Customize the plot
plt.title("Employees who got a promotion recently and spent <1 year with their current manager versus employees who have spent >=1 year.")
plt.xlabel("Years with Current Manager")
plt.ylabel("Attrition Rate")

# Show the plot
plt.show()
```
