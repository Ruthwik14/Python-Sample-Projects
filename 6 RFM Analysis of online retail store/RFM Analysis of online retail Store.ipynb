RFM (Recency Frequency Monetary) Analysis:-
RFM is a method used for analyzing customer value. It is commonly used in database marketing and direct marketing and has received particular attention in retail and professional services industries

RFM stands for the three dimensions:

Recency – How recently did the customer purchase?
Frequency – How often do they purchase?
Monetary Value – How much do they spend?

```python
#importing all important package..
import pandas as pd
import numpy as np
import datetime
```

```python
#load data into pandas dataframe..
df = pd.read_csv("Dataset.csv")
df.head()
```

```python
#information of dataset..
df.info()
```

```python
df['Price'] = df['Price'].str.replace(',','.')
df.head()
```

```python
# change datatype of price column
df = df.astype({"Price":"float64"})
df.info()
```

```python
# change datatype of InvoiceDate column
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')
df.info()
```

```python
# change datatype of Customer ID column
df['Customer ID'] = df['Customer ID'].astype(str).apply(lambda x: x.replace('.0',''))
df.info()
```

```python
# rename the column name for Customer ID to Customer_ID for future operation
df.rename(columns = {'Customer ID':'Customer_ID'}, inplace = True)
df.head()
```

```python
# creating new column total revenue as Tot_rev
df["Tot_Rev"]=(df['Price']*df['Quantity'])
df.head()
```

```python
#sorting the Customer_ID 
df = df.sort_values(by="Customer_ID",ascending=0)
df.head()
```

```python
#removing Not a Number rows in dataframe
df=df.dropna()
df.head()
```

```python
# In customer column their are un-related values need to removed
df.drop(df[df['Customer_ID'] == 'nan'].index, inplace = True)
df.head()
```

```python
#create empty data frame
RFM = pd.DataFrame()
print(RFM)
```

```python
# calculating no of time  purchase made by each customer
RFM['Freq'] = df.groupby("Customer_ID").StockCode.count()
RFM.head()
```

```python
# calculating Total revenue purchased by each customer
RFM['Tot_Rev'] = df.groupby("Customer_ID").Tot_Rev.sum()
RFM.head()
```

```python
# calculating Average revenue purchased by each customer
RFM['Avg_Rev'] = RFM['Tot_Rev']/RFM['Freq']
RFM.head()
```

```python
#sorting by Customer_ID 
RFM = RFM.sort_values(by="Customer_ID",ascending=0)
RFM.head()
```

```python
RFM = RFM.dropna()
RFM.head()
```

```python
#Creating the column called order_date(which will taken has todays date)
RFM['order_date'] = '1.1.2012 00:00'
```

```python
# change datatype of order_date column
RFM['order_date']  = pd.to_datetime(RFM['order_date'] , errors='coerce')
RFM.info()
```

```python
# calulating recent date purchansed by each cust
RFM['MAX_InvoiceDate'] = df.groupby("Customer_ID").InvoiceDate.max()
RFM.head()
```

```python
# calulating duration from last purchased 
RFM['duration'] =  RFM['order_date'] - RFM['MAX_InvoiceDate'] 
RFM.head()
```

```python
RFM = RFM.drop(['order_date','MAX_InvoiceDate'], axis=1)
RFM.head()
```

```python
# this is RFM (Recency Frequency Monetary) table for all customers
RFM = RFM.reset_index()
RFM.head()
```

```python
#creating 10 different buckets for duration (Deciles)
RFM['Recency_Deciles'] = pd.qcut(RFM['duration'],10)
RFM.head()
```

```python
#create empty dataframe called vintage
vin = pd.DataFrame()
print(vin)
```

```python
# calulating total revenue for each bucket
vin["Tot_rev"] = RFM.groupby('Recency_Deciles').Tot_Rev.sum()
vin.head(10)
```

```python
# calculating cumulative revenue
vin["CumSum_Rev"] = vin["Tot_rev"].cumsum()
vin.head(10)
```

```python
# calculating total revenue
vin["Tot_rev_Across_Deciles"] = RFM.Tot_Rev.sum()
vin.head(10)
```

```python
# calculating percentage cumilative revenue with respect to total revenue for all the customers
vin["perc_tot_rev"] = vin["CumSum_Rev"]/vin["Tot_rev_Across_Deciles"]
vin.head(10)
```

```python
#creating 10 different buckets wr.t frequency column
RFM['Feq_Deciles'] = pd.qcut(RFM['Freq'],10)
RFM.head()
```

```python
# create empty dataframe
rec = pd.DataFrame()
print(rec)
```

```python
# calculating total revenue for each bucket
rec["Tot_rev"] = RFM.groupby('Feq_Deciles').Tot_Rev.sum()
rec.head(10)
```

```python
# calculating cumulative revenue
rec["CumSum_Rev"] = rec["Tot_rev"].cumsum()
rec.head(10)
```

```python
# calculating total revenue for each bucket
rec["Tot_rev_Across_Deciles"] = RFM.Tot_Rev.sum()
rec.head(10)
```

```python
#calculating percentage total revenue with respect to total revenue of all the customers
rec["perc_tot_rev"] = rec["CumSum_Rev"]/rec["Tot_rev_Across_Deciles"]
vin.head(10)
```
